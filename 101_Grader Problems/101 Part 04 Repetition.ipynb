### Task 04_01 : Average (1 Star)


```python
# Key : รับค่าเเรกก่อน เเล้วค่อยวน while loop เพื่อเช็คตัวอื่น

# 1 : รับ input + define initial var.
num = input("Enter Number : ") #int (รับเข้าเป็น int)
total = 0
count = 0 

# 2 : check condition ว่าถ้าป้อน 'q' เเรก
if num == 'q' :
    count += 1 # เพิ่ม count เพื่อให้ไม่หาร 0
    print('No Data') # in case ค่าเเรก = 'q'
    
# 3 : while loop
while num != 'q' : ## Logic : รวมค่า -> + count -> รับค่า
    # 3.1 : รวมค่า numเดิม เข้าไป ใน total
    total += float(num)
    # 3.2 : เพิ่มค่า count +1 
    count += 1
    # 3.3 : รับค่า num เพื่อนำไปบวก
    num = input('Enter Another Value : ') #int or str ก็ได้ เพราะต้องการเช็คว่าเป็น 'q' มั้ย?

# 4 : เมื่อออก loop => หา average (ในเงื่อนไขที่ average > 0)
if (total / count) > 0 :
    average = total / count
    print( round(average,2) )
```

### Task 04_02 : Bisection Log10 (1.5 Star)


```python
## Logic : Key ของการหา bisection คือ หา inverse ของค่าที่ต้องการหา
        # Ex : อยากหา √a -> ต้องเช็คด้วย x^2
        # Ex : อยากหา ln x -> ต้องเช็คด้วย e^x

# 1 : กำหนดค่าเริ่มต้น
a = float(input('Enter Number for log10 : '))
lower = 0
upper = a
x = (lower + upper) / 2 

# 2 : กำหนดเงื่อนไข
while abs(a-(10**x)) > 1e-10 * max(a,(10**x)) : # กำหนดเงื่อนไขตรงข้ามค่าเข้าใกล้ (เพื่อให้ยังทำงานอยู่)
    
    # 2.1 : 10^x > a => เปลี่ยน upper ให้มีค่า x (ตัดฝั่งขวา)
    if 10**x > a :
        upper = x 
        
    # 2.2 : 10^x <= a => เปลี่ยนlower ให้มีค่า x (ตัดฝั่งซ้าย)
    else : 
        lower = x
        
    # 2.3 : เปลี่ยนค่า x ทุกครั้งที่ผ่านการ loop
    x = (lower + upper) / 2
    
# 3 : output x 6 ตน.
print( round(x,6) ) # ค่า log10 ของค่า input (Error ที่ค่า 616)
```

### Task 04_05 : Count Word (1 Star)


```python
# Note : นำ Logic ข้อ 04_04 มาใช้ต่อ (ตัด str ตัวเเปลกๆ)

## Logic
# 1. ตัด str เเปลกๆออก (for loop [string] เช็ค) -> เพิ่ม str ว่าง => ถ้าเเปลก ให้เปลี่ยนเป็น spacebar , ถ้าไม่เเปลก คืน strdaidai (ตัวที่กำลังเช็ค)
# 2. split str_all ให้เป็น list (of str) => ใช้ for loop เช็คต่อ

# 1 : รับ str_want กับ str_check เข้ามา + define ตป. count
str_check = str(input('Enter Check Word : '))
str_all = input('Enter String : ') # str
count = 0

# 2 : สร้าง str เปล่า เพื่อนำไปเเทน str เเปลกๆ -> สร้าง for loop เพื่อตัดตัวอักษร
str_replace = ''
for strdaidai in str_all : # strdaidai = ตน.ใดๆ ใน str
    
    # ถ้ามีตัวเเปลกๆ ให้เปลี่ยนค่าเป็น spacebar ( )
    if strdaidai in '!@"\'[]#()[].' : 
        str_replace += ' '
        
    # ถ้าไม่เเปลก ให้คืน strdaidai
    else : 
        str_replace += strdaidai

## ตอนนี้ str_all => ถูก replace เป็น str_replace เเล้ว

# 3 : split str_all ให้เป็น list (of str) -> for loop เช็คจำนวนคำ
str_split = str_replace.split() ## ตัด spacebar ทุกตัวออก

# 4 : for loop (string) เพื่อเช็ค จน.คำ
for element in str_split : # check element in split
    if str_check == element : # if str เป็นตัวเดียวกับstr ใน list
        count += 1
        
# 4 : print output
print(count)
```

### Task 04_06 : Print Triangle (2 Star)


```python
## Logic : ช่องที่ไม่เเสดงคือ spacebar => เเต่ละบรรทัด มีจน. 2h-1 ตัว มี h บรรทัด

## จะได้ว่า pattern บรรทัดที่ n (ไม่เอา บรรทัดเเรกกับสุดท้าย)  : (.) h-n ตัว + (*) 1 ตัว + (.) 2(n-1)-1 ตัว + (*) 1 ตัว + spacebar h-n ตัว 
# Ex : h = 8 , n = ถ                                 : (.) 8-5 = 3 ตัว + (*) 1 ตัว + (.) 2(5-1) - 1 = 7 ตัว + (*) 1 ตัว + spacebar 8-5 = 3 ตัว

# 1 : รับ input height (int)
h = int(input('Enter Height : '))

# 2 : วน for loop ตาม จน.บรรทัดที่กรอกเข้ามา

for n in range(1,h+1) : # เริ่ม 1 ถึง h (ไม่เอา h+1)
    ## สร้าง 3 เงื่อนไข : 1.บรรทัดเเรก 2.บรรทัดที่เหลือ 3.บรรทัดสุดท้าย
    
    # บรรทัดเเรก : n = 1 => (.) h-1 ตัว (*) 1 ตัว ( ) h-1 ตัว
    if n == 1 :
        print( '.' * (h-1) + '*' * 1  + ' ' * (h-1) )
    
    # บรรทัดสุดท้าย : n = h => (*) 2h-1 ตัว
    elif n == h :
        print( '*' * (2*h - 1) ) 

    # บรรทัดอื่นๆ : n = n => (.) h-n ตัว + (*) 1 ตัว + (.) 2(n-1)-1 ตัว + (*) 1 ตัว + spacebar h-n ตัว
    else :
        print('.' * (h-n) + '*' * 1 + '.' * (2*(n-1)-1) + '*' * 1 + ' ' * (h-n) ) 
```

### Task 04_07 : Bisection Log10 (Ver.2) (2 Star)


```python
## จาก 04_02 Bisection Log10 : Error ที่ค่า 616

# Part 1 : เเปลงค่า U (เข้า while loop ก่อน)
a = float(input('Enter A : ')) ##!!ระวัง!! a รับเป็น float ไว้ก่อน เซฟกว่า
count = 0 # จน.ครั้งในการ mod ให้เหลือ 0 (ที่จริงก็คือ จน.หลักอะเเหละ)

## เจอปัญหาว่า a ถูกเปลี่ยนค่าเป็น 0 -> ไม่เข้า while => เเก้โดยเพิ่มตัวเเปรไป
a_old = a 

while a != 0 : # ทำจนกว่า a = 0
    a = (a) // 10
    count += 1   
# ตอนนี้ ได้ count เเล้ว (a ไม่ใช้ต่อ)
    
# Part 2 : Bisection Log10

# 2.1 : define initial variable (ระวัง! เอาตัวเเปร a ก่อนถูกเปลี่ยนค่ามาใช้)
lower = 0
upper = count  #int
x = (lower + upper)/2 #float

# 2.2 : while loop 
    # ==> Logic : หา inverse ของ log 10 = 10^x = b , เเล้วตั้งเงื่อนไขตรงข้าม เพื่อยังให้ while ทำงานอยู่ 
while abs(a_old - 10**x) > ((1e-10) * max(a_old,10**x)) :
    
    # 2.2.1 : ถ้า 10^x > a => เปลี่ยน upper เป็น x (ตัดขวา)
    if 10**x > a_old :
        upper = x
    # 2.2.2 : ถ้า 10^x < a => เปลี่ยน lower เป็น x (ตัดซ้าย)
    elif 10**x < a_old :
        lower = x
        
    # 2.2.3 : เปลี่ยนค่า x ทุกครั้ง
    x = (lower + upper) /2
        
# 3 : print output (x) 6 ตน.
print( round(x,6) )
```

### Task 04_08 (12) : Run Length Encoding (RLE) (2 Star)


```python
# Logic : สร้าง ตป. check เเรก -> ถ้าไม่เหมือนเดิม ให้ print ค่าเดิม + count เเล้วเปลี่ยนค่าใหม่ (อย่าลืม print ค่าท้ายตอนจบ loop)

# 1 : รับ input rle เเล้วไล่เช็คทีละตัว
rle = str(input('Enter RLEs : ')) # str

# 2 : กำหนดตัวเเปรที่ต้องใช้ (count , check , rle_output)
count = 1 #  กำหนดตัวเเรกเป็น 1
check = rle[0] # ให้ตัวเเรกที่เช็คเป็น index 0
rle_output = '' # str ว่าง เพื่อค่อยๆเพิ่มค่าเข้าไป

# 3 : for loop (range) วนเช็คเเต่ละตัว !! กำหนด range เริ่มเป็น 1 (ไม่ต้องเช็ค index 0 = ตัวเเรก) ถึง len (ทำเเค่ len - 1 รอบ)
for i in range(1 , len(rle)) :
    
    # 3.1 : cd1 -> check ตัวเดียวกับ rle[i] => count +1
    if check == rle[i] :
        count += 1
        
    # 3.2 : cd2 -> check คนละตัวกับ rle[i] => เเสดง check , count เดิมก่อน เเล้วค่อยเปลี่ยนค่า
    elif check != rle[i] :
        # เพิ่มค่าเดิมเข้าไปใน output ก่อน
        rle_output += (str(check) + ' ' + str(count) + ' ') # เพิ่ม count เดิม & check เดิม เข้าไปในoutput
        # เปลี่ยนค่า =. count = 1 , check = rle[i] = ตัวใหม่ที่ไม่เหมือน
        count = 1
        check = rle[i]
        
# 4 : จบ loop เเล้ว -> อย่าลืมเช็คตัวสุดท้าย
    # 4.1 : เพิ่ม count , check ของตัวสุดท้าย เข้าไปใน rle_output ก่อน
rle_output += (str(check) + ' ' + str(count) + ' ') # ค่า count , check ของตัวสุดท้าย
    # 4.2 : print ผลสุดท้ายของ rle_output
print(rle_output) #str
```

### Task 04_09 (12) : Zigzag Ver.1 (2 Star)


```python
# โกงเเม่ม ใช้ list กับ min,max ==> ลองไม่ใช้ ใน 04_10 Zigzag (Ver.2)

# 1 : รับ loop + กำหนด ตป.ที่จำเป็น
loop = int(input('Enter loop : ')) # รับ จน.loop
red , blue = [] , [] # สร้าง list ว่าง สายเเดง & สายน้ำเงิน

# 2 : วน for loop เพื่อรับ x,y เข้ามาเรื่อยๆ -> append ลง list red & blue
for i in range(loop) :
    x, y = input('Enter X , Y : ').split()
    
    # check index คู่ - คี่ เพื่อ append สลับกัน
    if i % 2 == 0 : # index คู่
        red.append(int(x))
        blue.append(int(y))
    elif i % 2 != 0 : # index คี่
        red.append(int(y))
        blue.append(int(x))   
# ตอนนี้ ได้ list (of int) red กับ blue เเล้ว

# 3 : รับ algo เข้ามา เพื่อเช็คว่าเป็นเเบบ Zag-Zig หรือ Zig-Zag
algo = input('Enter Zig-Zag or Zag-Zig : ')

    # 3.1 : ถ้าเป็น Zig-Zag ==> หา min_red กับ max_blue โดยใช้ max() , min()
if algo == 'Zig-Zag' :
    min_red = min(red)
    max_blue = max(blue)
    print(min_red , max_blue)
    
    # 3.2 : ถ้าเป็น Zig-Zag ==> สลับ list red,blue ก่อน ==> หา min_red กับ max_blue โดยใช้ max() , min(
elif algo == 'Zag-Zig' :
    red , blue = blue , red # สลับ list กัน
    min_red = min(red)
    max_blue = max(blue)
    print(min_red , max_blue)
```

### Task 04_10 (20) : Zigzag Ver.2 (3 Star)


```python
# 1 : input -> split & give first nuber to max&min
info = input() # first line
i = 0
x_min = int(info.split()[0])
y_max = int(info.split()[1])

# Case 1 : Zig-Zag
while (info != 'Zig-Zag') and (info != 'Zag-Zig') :
    # case 1 : loop = 0
    if i == 0 :
        info = info.split()
        x_min = int(info[0]); x_max = int(info[0])
        y_max = int(info[1]); y_min = int(info[1])
    # case 2 : loop even
    elif (i % 2 == 0) and (i != 0) :
        info = info.split()
        if int(info[0]) < x_min : x_min = int(info[0])
        if int(info[1]) > y_max : y_max = int(info[1])
        # we can use if because x_max doesnt matter on x_min
        if int(info[0]) > x_max : x_max = int(info[0])
        if int(info[1]) < y_min : y_min = int(info[1])
    # case 3 : loop odd    
    elif (i % 2 == 1) :
        info = info.split()
        if int(info[1]) < x_min : x_min = int(info[1])
        if int(info[0]) > y_max : y_max = int(info[0])
        # we can use if because x_max doesnt matter on x_min
        if int(info[1]) > x_max : x_max = int(info[1])
        if int(info[0]) < y_min : y_min = int(info[0])
    # always plus loop by 1
    i += 1
    info = input()
    
# check info : show min , max (-71 , 80 -> y_min , x_max)
if info == 'Zig-Zag' : 
    print(x_min , y_max)
elif info == 'Zag-Zig' :
    print(y_min , x_max)
```

### Task 04_12 (002) : Partition (Flowchart) (2 Star)


```python
# 1 : รับ input คั่นด้วยช่องว่าง
d = [int(e) for e in input().split()] # str -> list (of int)

# 2 : ทำ step 2
p = d[-1] # index ตัวสุดท้ายของ d
i = -1
j = 0
n = len(d) # จน.ข้อมูลใน list

# 3 : ทำ while loop
while j < n-1 : ## สังเกตลูกศร ถ้า False จะออก loop ==> จะได้ว่า ถ้า True จะยังอยู่ใน loop
    
    # 3.1 : condition ที่ว่า d[j] <= p
    if d[j] <= p :
        i += 1
        d[i] , d[j] = d[j] , d[i] # สลับค่า d[i] กับ d[j]
    else :
        pass
    
    # 3.2 : บังคับทำ j += 1 ไม่ว่าจะ if หรือ else
    j += 1
    
# 4 : ออก while loop -> ทำ step 4 ต่อ

# 4.1 : สลับค่า d[i+1] กับตัวขวาสุดของ d (d[-1])
d[i+1] , d[-1] = d[-1] , d[i+1] 

# 4.2 : print d
print(d)
```

### Task 04_13 : Multiple Choice Question (MCQ) (Function) (1 Star)


```python
# 1 : สร้าง fn โดยมี parameter : sol,ans เป็น str Ex : 'AAAAB' , 'BDCDA'
def grade_mcq(sol,ans) :
    
    count = 0 # define ตป. count ก่อน
    
    # 2 : สร้าง cd ดักไว้ (ความยาว str ไม่เท่ากัน)
    if len(sol) != len(ans) :
        return -1
    
    # 3 : สร้าง for loop (range = จน.ความยาว str)
    for i in range(len(ans)) : # กำหนด range เป็น len ของ ans
        if ans[i] == sol[i] :
            count += 1
    
    # ออก for loop => อย่าลืม return count
    return count
        
# 4 : execute input เพื่อ check printed fn
exec(input())
```

### Task 04_14 : Print Triangle (Function) (2 Star)


```python
# Logic : print ค่า ความยาว 2h-1 สูง 2h-1 (ที่ไมมีตัวคือช่องว่าง)
# ประเด็นคือ มี 3 condition ที่ต่างกัน

# (สังเกต pattern) ==> n = จน.บรรทัด
# บรรทัดเเรก : (.) h-1 ตัว , (*) 1 ตัว , ( ) h-1 ตัว
# บรรทัดอื่นๆ : (.) h-n ตัว , (*) 1 ตัว , (.) 2(n-1)-1 ตัว , (*) 1 ตัว , ( ) h-n ตัว
# บรรทัดสุดท้าย : (*) 2h-1 ตัว

# 1 : สร้าง fn โดยมี parameter h เป็นความสูง 3 เหลี่ยม ==> วน for loop จน.บรรทัด
def print_triangle(h) :
    for n in range(1,h+1) : # เริ่มบรรทัด 1 (n=1) , จบที่ h
        
        if n == 1 : # บรรทัดเเรก
            print( '.' * (h-1) + '*' * 1 + ' ' * (h-1) )
            
        elif n == h : # บรรทัดสุดท้าย
            print( '*' * (2*h - 1) )
            
        else : # บรรทัดอื่นๆ
            print( '.' * (h-n) + '*' * 1 + '.' * (2*n-3) + '*' * 1 + ' ' * (h-n) )
            
# 2 : execute input
exec(input())
```

### Task 04_15 : Run Length Encoding (Function) (2 Star)


```python
# Key : เช็คค่าไปเรื่อย (set ค่าเริ่มต้นตัวเเรก) -> ถ้าไม่เหมือน ให้เปลี่ยนตัวเช็ค & เพิ่มค่าเดิม

# 1 : สร้าง fn RLE : return เป็น list (of list)
def RLE (t) :
    
    # 2 : check ก่อนว่า t ที่เป็น parameter เป็น list ว่าง มั้ย?
    if t == '' :
        return [] # return str ว่าง
    
    # 3 : กำหนดตัวเเปรที่ต้องใช้ก่อน
    count = 1 
    check = t[0]
    rle_output = []
    
    # 4 : for loop เช็ค rle => เริ่มค่า index 1 ถึง len (ทำ len-1 รอบ)
    for i in range(1 , len(t)) :
        # cd1 : ถ้าค่าเหมือนเดิม -> + count
        if check == t[i] :
            count += 1
        # cd2 : ถ้าค่าไม่เหมือนเดิม -> เพิ่มค่าเดิมลง list (append list) เเล้วค่อยเปลี่ยนค่า check , count
        elif check != t[i] :
            # append list
            rle_output.append( [str(check),count] ) # append list
            # เปลี่ยนค่า check , count
            count = 1
            check = t[i] 
            
    # 5 : จบ for loop เเล้ว -> เพิ่มค่า check , count ตัวสุดท้าย 
    rle_output.append( [str(check),count] ) # append list
    
    # 6 : return rle_output
    return rle_output

# 7 : exec input เพื่อ check fn
exec(input())
```

### Task 04_16 : Partylist Member (3 Star)




```python
# 1 : input first info (use split) + define vacant list
info = input()
party = [] ; votes = [] ; sum_vote = 0

# 2 : Loop 1 : add ele in list -> create 3 list
while info != 'END' :
    party += [ info.split()[0] ] # .append() is better but in grader#1 recommend this one
    votes += [ int(info.split()[1]) ] # .append() is better but in grader#1 recommend this one
    sum_vote += int(info.split()[1]) #sum each votes to sum_votes
    info = input()

# now we have list : party (list of str) , votes (list of int) , sum_votes (int)

# 3 : Loop 2 : find avg votes per ps / create int member list & partial member list
avg = sum_vote / 100
partylist = [0] * len(votes);  sum_member = 0;  partial_member = []
for i in range(len(votes)) :
    partylist[i] = votes[i] / avg # find #partylist_mb of each person
    sum_member += (votes[i] // avg) 
    partial_member += [ partylist[i] % 1 ] # mod 1 -> we can get 0.x people for ceiling
    partylist[i] = int(partylist[i]) # mod to int at last

# 4 : Loop 3 : ceiling partial member
for j in range(100 - int(sum_member)) : # loop left from 100
    # 4.1 : index & ceil
    ceiling = max(partial_member)
    id = partial_member.index(ceiling)
    partylist[id] = partylist[id] + 1
        # cuz we already change to list (of int) -> so we can +1 directly
    partial_member[id] = 0 # change that index to 0 -> compare in list
    
    # check if sum(partylist) is 100 -> break
    if sum(partylist) == 100 :
        break

# 5 : Loop 4 : add member in list of list for sorting & print
info = []
for i in range(len(partylist)) :
    if partylist[i] > 0 : # append member for #partylist > 0 only
        info += [ [partylist[i] , votes[i] , party[i]] ] # + list (of list)
        
# 6 : reverse sorting (descending) -> loop for printing
info.sort(reverse = True)
for j in range(len(info)) :
        print(info[j][2] , info[j][0] , info[j][1])
```
