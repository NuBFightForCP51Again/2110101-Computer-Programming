```python
# Key : ‡∏õ‡πâ‡∏≠‡∏ô‡∏Ñ‡πà‡∏≤‡πÄ‡∏Ç‡πâ‡∏≤‡∏°‡∏≤ 12 ‡∏´‡∏•‡∏±‡∏Å ‡πÄ‡∏û‡∏∑‡πà‡∏≠‡πÄ‡∏ä‡πá‡∏Ñ‡∏´‡∏•‡∏±‡∏Å‡∏ó‡∏µ‡πà 13

# Note : ‡∏£‡∏´‡∏±‡∏™‡∏ö‡∏±‡∏ï‡∏£‡∏õ‡∏£‡∏∞‡∏ä‡∏≤‡∏ä‡∏ô 1 - 1  0  2  2 - 0  0  2  2  4 - 7   3 - ??? (‡∏Ç‡∏≠‡∏á‡∏ï‡∏±‡∏ß‡πÄ‡∏≠‡∏á)
# ‡πÄ‡πÄ‡∏ó‡∏ô‡∏ï‡∏±‡∏ß‡πÄ‡πÄ‡∏õ‡∏£‡∏´‡∏•‡∏±‡∏Å‡∏ó‡∏µ‡πà        n0 n1 n2 n3 n4  n5 n6 n7 n8 n9  n10 n11 n12

# Sol1 : ‡∏ß‡∏¥‡∏ò‡∏µ‡∏ï‡∏£‡∏á 

# 1 : ‡∏£‡∏±‡∏ö Citizen ID 12 ‡∏´‡∏•‡∏±‡∏Å‡πÄ‡∏Ç‡πâ‡∏≤‡∏°‡∏≤ (str) [‡∏£‡∏π‡πâ‡∏•‡πà‡∏ß‡∏á‡∏´‡∏ô‡πâ‡∏≤ ‡∏≠‡∏¢‡πà‡∏≤‡πÉ‡∏ä‡πâ‡∏ä‡∏∑‡πà‡∏≠ ‡∏ï‡∏õ.‡∏ß‡πà‡∏≤ id]
cid = input('Enter 12 digit : ') #str 12 ‡∏´‡∏•‡∏±‡∏Å 

# 2 : index ‡πÄ‡πÄ‡∏ï‡πà‡∏•‡∏∞‡∏ï‡∏±‡∏ß ‡πÄ‡πÄ‡∏•‡πâ‡∏ß‡πÄ‡πÄ‡∏õ‡∏•‡∏á‡πÄ‡∏õ‡πá‡∏ô int ‡πÄ‡∏•‡∏¢ -> ‡πÄ‡∏û‡∏∑‡πà‡∏≠‡∏ô‡∏≥‡πÑ‡∏õ‡∏Ñ‡∏≥‡∏ô‡∏ß‡∏ì
n0 = int(cid[0]); n1 = int(cid[1]); n2 = int(cid[2]); n3 = int(cid[3])
n4 = int(cid[4]); n5 = int(cid[5]); n6 = int(cid[6]); n7 = int(cid[7])
n8 = int(cid[8]); n9 = int(cid[9]); n10 = int(cid[10]); n11 = int(cid[11])

# 3 : ‡∏ô‡∏≥‡∏°‡∏≤‡∏Ñ‡∏≥‡∏ô‡∏ß‡∏ì‡∏´‡∏≤ n12 ‡∏ï‡πà‡∏≠

# 3.1 : ‡∏™‡∏°‡∏°‡∏ï‡∏¥‡πÉ‡∏´‡πâ‡∏Å‡πâ‡∏≠‡∏ô 13n0 + ... + 2n11 ‡πÄ‡∏õ‡πá‡∏ô x
x = 13*n0 + 12*n1 + 11*n2 + 10*n3 + 9*n4 + 8*n5 + 7*n6 + 6*n7 + 5*n8 + 4*n9 + 3*n10 + 2*n11

# 3.2 : ‡πÄ‡∏Ç‡πâ‡∏≤‡∏™‡∏π‡∏ï‡∏£ ‡∏´‡∏≤ n12
n12 = (11 - (x % 11)) % 10 #float

# 4 : print ‡∏ï‡∏≤‡∏° format
print(cid[0], cid[1:5] , cid[5:10] , cid[10:12] , str(int(n12)) )

# ‡πÄ‡∏£‡∏≤‡∏ó‡∏≥‡∏Å‡∏≤‡∏£ slicing index ‡πÄ‡∏≠‡∏≤‡∏ä‡πà‡∏ß‡∏á‡∏ó‡∏µ‡πà‡πÄ‡∏£‡∏≤‡∏™‡∏ô‡πÉ‡∏à‡∏≠‡∏≠‡∏Å‡∏°‡∏≤

#index 0 = '1'
#index 1-4 = '1022'
#index 5-9 = '00224'
#index 10-11 = '73'
#index ‡∏´‡∏•‡∏±‡∏Å‡∏™‡∏∏‡∏î‡∏ó‡πâ‡∏≤‡∏¢ ‡∏à‡∏≤‡∏Å‡∏™‡∏π‡∏ï‡∏£ ‡∏à‡∏∞‡πÑ‡∏î‡πâ‡πÄ‡∏õ‡πá‡∏ô '4'

### ‡∏ö‡∏£‡∏£‡∏ó‡∏±‡∏î‡πÄ‡∏¢‡∏≠‡∏∞ ‡πÄ‡∏£‡∏≤‡∏¢‡πà‡∏≠‡πÑ‡∏î‡πâ ‡∏î‡πâ‡∏ß‡∏¢ Sol2
```

```python
# Sol 2 : ‡∏™‡∏±‡∏á‡πÄ‡∏Å‡∏ï‡∏´‡∏ô‡πà‡∏≠‡∏¢ ‡∏à‡∏∞‡πÄ‡∏´‡πá‡∏ô‡∏ö‡∏≤‡∏á‡∏≠‡∏¢‡πà‡∏≤‡∏á
# ‡∏•‡∏≠‡∏á‡∏™‡∏°‡∏°‡∏ï‡∏¥ ‡∏ï‡∏õ.‡πÄ‡∏£‡∏¥‡πà‡∏°‡∏ï‡πâ‡∏ô‡πÄ‡∏õ‡πá‡∏ô i

# 1 : ‡∏£‡∏±‡∏ö‡∏Ñ‡πà‡∏≤ str 12 ‡∏´‡∏•‡∏±‡∏Å + define ‡∏ï‡∏õ.‡πÄ‡∏£‡∏¥‡πà‡∏°‡∏ï‡πâ‡∏ô
cid = str(input('Enter Citizen ID : ')) # str 12 ‡∏´‡∏•‡∏±‡∏Å
i = 0

# 2 : ‡πÄ‡∏û‡∏¥‡πà‡∏°‡∏Ñ‡πà‡∏≤‡πÄ‡πÄ‡∏ó‡∏ô ‡∏ï‡∏õ.‡πÄ‡∏î‡∏¥‡∏° ‡πÇ‡∏î‡∏¢‡∏Å‡∏≤‡∏£ += ‡πÑ‡∏õ‡πÄ‡∏£‡∏∑‡πà‡∏≠‡∏¢ ‡πÇ‡∏î‡∏¢‡πÄ‡∏û‡∏¥‡πà‡∏°‡∏Ñ‡πà‡∏≤ i ‡∏ó‡∏µ‡∏•‡∏∞ 1 ‡∏ó‡∏∏‡∏Å‡∏Ñ‡∏£‡∏±‡πâ‡∏á (‡∏ù‡∏∂‡∏Å‡∏™‡∏±‡∏á‡πÄ‡∏Å‡∏ï‡∏î‡∏π)
s = 13 * int(cid[i]);        i += 1     # i = 1
s += (13-i) * int(cid[i]);   i += 1     # i = 2 
s += (13-i) * int(cid[i]);   i += 1     # i = 3
s += (13-i) * int(cid[i]);   i += 1     # i = 4
s += (13-i) * int(cid[i]);   i += 1     # i = 5 
s += (13-i) * int(cid[i]);   i += 1     # i = 6
s += (13-i) * int(cid[i]);   i += 1     # i = 7
s += (13-i) * int(cid[i]);   i += 1     # i = 8 
s += (13-i) * int(cid[i]);   i += 1     # i = 9
s += (13-i) * int(cid[i]);   i += 1     # i = 10
s += (13-i) * int(cid[i]);   i += 1     # i = 11
s += (13-i) * int(cid[i]);   i += 1     # i = 12

n12 = (11 - (s % 11)) % 10

# 3 : slice ‡πÄ‡πÄ‡∏•‡πâ‡∏ß print ‡∏ï‡∏≤‡∏° format
print(cid[0], cid[1:5] , cid[5:10] , cid[10:12] , str(n12) ) 
# index   0       1-4       5-9        10-11            12


### Challenge  ‡∏•‡∏≠‡∏á‡πÄ‡∏Ç‡∏µ‡∏¢‡∏ô‡πÇ‡∏î‡∏¢‡πÉ‡∏ä‡πâ repetition ‡∏î‡∏π
```

```python
# Sol3 : ‡∏•‡∏≠‡∏á‡πÉ‡∏ä‡πâ while ‡∏î‡∏π

# 1 : ‡∏£‡∏±‡∏ö input str 12 ‡∏´‡∏•‡∏±‡∏Å + ‡∏Å‡∏≥‡∏´‡∏ô‡∏î ‡∏ï‡∏õ. loop (i) +  ‡∏ï‡∏õ. sum (‡πÄ‡∏û‡∏∑‡πà‡∏≠‡∏£‡∏ß‡∏°‡∏Ñ‡πà‡∏≤‡∏Ç‡∏≠‡∏á‡πÄ‡πÄ‡∏ï‡πà‡∏•‡∏∞‡∏´‡∏•‡∏±‡∏Å‡∏ï‡∏≤‡∏°‡∏™‡∏π‡∏ï‡∏£)
cid = str(input('Enter CItizen ID : '))
i = 0
sum = 0

# 2 : ‡πÄ‡∏Ç‡πâ‡∏≤ while loop ‡πÄ‡∏û‡∏∑‡πà‡∏≠‡πÄ‡∏û‡∏¥‡πà‡∏°‡∏Ñ‡πà‡∏≤ s ‡πÑ‡∏õ‡πÄ‡∏£‡∏∑‡πà‡∏≠‡∏¢‡πÜ

while i < len(cid) : # ‡∏ï‡∏£‡∏≤‡∏ö‡πÉ‡∏î‡∏ó‡∏µ‡πà‡∏¢‡∏±‡∏á‡πÑ‡∏°‡πà‡πÄ‡∏Å‡∏¥‡∏ô‡∏´‡∏•‡∏±‡∏Å‡∏™‡∏∏‡∏î‡∏ó‡πâ‡∏≤‡∏¢
    sum += (13-i) * (int(cid[i])) # ‡πÄ‡∏Ç‡πâ‡∏≤‡∏™‡∏π‡∏ï‡∏£ (‡∏≠‡∏¢‡πà‡∏≤‡∏•‡∏∑‡∏°‡πÄ‡πÄ‡∏õ‡∏•‡∏á‡∏Ñ‡πà‡∏≤‡πÄ‡πÄ‡∏ï‡πà‡∏•‡∏∞‡∏´‡∏•‡∏±‡∏Å‡πÄ‡∏õ‡πá‡∏ô int)
    i += 1 # ‡πÄ‡∏û‡∏¥‡πà‡∏°‡∏Ñ‡πà‡∏≤‡∏Ç‡∏∂‡πâ‡∏ô‡πÑ‡∏õ‡∏ó‡∏µ‡∏•‡∏∞ 1
    
n12 = (11 - (sum % 11)) % 10
print(cid[0], cid[1:5] , cid[5:10] , cid[10:12] , str(n12) ) 
```

### Task 02_04 : N-Digits (1 Star)


```python
# Logic : ‡∏Å‡∏£‡∏≠‡∏Å‡πÄ‡∏•‡∏Ç ‡∏Ñ‡πà‡∏≤ & ‡∏´‡∏•‡∏±‡∏Å -> ‡∏ñ‡πâ‡∏≤ ‡∏à‡∏ô.‡∏´‡∏•‡∏±‡∏Å ‡∏°‡∏≤‡∏Å‡∏Å‡∏ß‡πà‡∏≤ ‡πÉ‡∏´‡πâ‡πÄ‡∏õ‡πá‡∏ô format ‡∏ï‡∏≤‡∏° ‡∏à‡∏ô.‡∏´‡∏•‡∏±‡∏Å -> ‡∏ñ‡πâ‡∏≤‡∏ô‡πâ‡∏≠‡∏¢‡∏Å‡∏ß‡πà‡∏≤ ‡πÉ‡∏´‡πâ‡∏Å‡∏£‡∏≠‡∏Å‡∏Ñ‡πà‡∏≤‡πÄ‡∏î‡∏¥‡∏°

# Step 1 : input : ‡∏£‡∏±‡∏ö input M,N (int) ‡πÄ‡∏Ç‡πâ‡∏≤‡∏°‡∏≤ ‡πÇ‡∏î‡∏¢ M ‡πÄ‡πÄ‡∏ó‡∏ô‡∏Ñ‡πà‡∏≤‡∏ï‡∏±‡∏ß‡πÄ‡∏•‡∏Ç , N ‡πÄ‡πÄ‡∏ó‡∏ô‡∏à‡∏ô.‡∏´‡∏•‡∏±‡∏Å
M = int(input()) #value (int) -> ‡πÇ‡∏à‡∏ó‡∏¢‡πå‡∏ö‡∏≠‡∏Å
N = int(input()) #digit (int) -> ‡πÇ‡∏à‡∏ó‡∏¢‡πå‡∏ö‡∏≠‡∏Å

# Step 2 : process : check condition ‡∏ß‡πà‡∏≤ digit ‡∏°‡∏≤‡∏Å‡∏Å‡∏ß‡πà‡∏≤ value ‡∏°‡∏±‡πâ‡∏¢? 

# define ‡∏ï‡∏õ. ‡∏à‡∏ô.‡∏´‡∏•‡∏±‡∏Å‡∏Ç‡∏≠‡∏á M ‡∏Ç‡∏∂‡πâ‡∏ô‡∏°‡∏≤
digits_M = len(str(M)) # Ex : len(str(12345 [int] )) = len('12345') = 5 [int]

# check condition ‡∏£‡∏∞‡∏´‡∏ß‡πà‡∏≤‡∏á digit ‡∏Å‡∏±‡∏ö value

# Case 1 : ‡∏ñ‡πâ‡∏≤ digit > value -> ‡πÉ‡∏™‡πà 0 ‡∏î‡πâ‡∏≤‡∏ô‡∏´‡∏ô‡πâ‡∏≤ value ‡∏ï‡∏≤‡∏°‡πÄ‡∏®‡∏©‡∏ó‡∏µ‡πà‡πÄ‡∏´‡∏•‡∏∑‡∏≠
if N > digits_M : # int > int
    # define ‡∏ï‡∏õ. ‡∏à‡∏ô. ‡πÄ‡∏•‡∏Ç 0 ‡∏Å‡πà‡∏≠‡∏ô‡∏´‡∏ô‡πâ‡∏≤ value
    zeros_b4_value = N - len(str(M)) # Ex : 5 - len('123') = 5-3 = 2
    print( '0' * zeros_b4_value + str(M) ) #Ex : '0' * 2 + str(12) = '0012'
    
# Case 2 : ‡∏ñ‡πâ‡∏≤ digit < value -> print ‡∏Ñ‡πà‡∏≤‡πÄ‡∏î‡∏¥‡∏° ‡∏≠‡∏≠‡∏Å‡∏°‡∏≤
else : 
    print(str(M)) # print ‡∏Ñ‡πà‡∏≤‡∏ó‡∏µ‡πà‡πÄ‡∏õ‡πá‡∏ô str ‡∏Ç‡∏≠‡∏á value ‡∏≠‡∏≠‡∏Å‡∏°‡∏≤

# ‡∏•‡∏≠‡∏á run ‡πÉ‡∏ô grader ‡πÄ‡πÄ‡∏•‡πâ‡∏ß -> 100 wa
```

### Solution 02_04 : N-Digits (1 Star)

```python
# 1 :  ‡∏£‡∏±‡∏ö input M,N
M = input('Enter Value : ')         # '12345'
N = int(input('Enter Digit : '))    # 5

# 2 : ‡∏´‡∏≤ ‡∏à‡∏ô.'0' ‡∏ó‡∏µ‡πà‡∏ï‡πâ‡∏≠‡∏á‡πÄ‡∏ï‡∏¥‡∏°
diff = N - len(M)
diff = max(diff,0) # for condition N < M

## ‡∏°‡∏µ‡∏õ‡∏±‡∏ç‡∏´‡∏≤‡∏ß‡πà‡∏≤ ‡∏ñ‡πâ‡∏≤ N (‡∏´‡∏•‡∏±‡∏Å) < M (‡∏Ñ‡πà‡∏≤) -> Ex : 2 - len('123') = -1 => diff ‡πÄ‡∏õ‡πá‡∏ô ‡∏•‡∏ö 
## ‡πÄ‡πÄ‡∏Å‡πâ‡πÑ‡∏î‡πâ ‡∏î‡πâ‡∏ß‡∏¢‡∏Å‡∏≤‡∏£ ‡πÉ‡∏ä‡πâ fn max ‡πÄ‡∏ó‡∏µ‡∏¢‡∏ö‡∏Å‡∏±‡∏ö‡∏Ñ‡πà‡∏≤ 0 -> ‡∏à‡∏∞‡πÑ‡∏î‡πâ‡πÄ‡∏õ‡πá‡∏ô '' ‡∏≠‡∏≠‡∏Å‡∏°‡∏≤‡∏≠‡∏¢‡∏π‡πà‡∏î‡∏µ

# 3 : output
print( '0' * diff + str(M))

### Note : str x ‡∏à‡∏ô.‡∏•‡∏ö = '' or [] -> ‡πÑ‡∏°‡πà‡∏Ñ‡∏∑‡∏ô‡∏≠‡∏≠‡∏Å‡∏°‡∏≤ ‡πÄ‡∏û‡∏£‡∏≤‡∏∞‡∏°‡∏≠‡∏á‡πÄ‡∏õ‡πá‡∏ô‡πÑ‡∏°‡πà‡∏°‡∏µ‡∏ï‡∏±‡∏ß‡∏Ñ‡∏π‡∏ì -> ‡πÑ‡∏î‡πâ‡∏Ñ‡πà‡∏≤‡πÄ‡∏ó‡πà‡∏≤‡πÄ‡∏î‡∏¥‡∏°
# Ex : -1 * 'hello' = ''
```

### Solution 02_05 : Weekly Sales (1 Star)

```python
# Step 1 : input : ‡∏£‡∏±‡∏ö input A,B ‡πÉ‡∏ô‡∏£‡∏π‡∏õ [x,y,z]
u_input = str(input()) #[x1,y1,z1]
v_input = str(input()) #[x2,y2,z2]

# Step 2 : process : ‡∏ï‡∏±‡∏î [] ‡∏≠‡∏≠‡∏Å‡∏Å‡πà‡∏≠‡∏ô ‡πÄ‡∏û‡∏∑‡πà‡∏≠‡∏à‡∏∞‡∏ó‡∏≥‡∏Å‡∏≤‡∏£ index => ‡πÉ‡∏ä‡πâ‡∏Ñ‡∏≥‡∏™‡∏±‡πà‡∏á strip('[]')
u_strip = u_input.strip('[]') # 'x1,y1,z1'
v_strip = v_input.strip('[]') # 'x2,y2,z2'

# Step 3 : process : split ‡πÄ‡∏û‡∏∑‡πà‡∏≠‡∏ó‡∏≥‡∏Å‡∏≤‡∏£‡πÄ‡πÄ‡∏¢‡∏Å ‡∏≠‡∏õ‡∏Å.‡∏Ç‡∏≠‡∏á vector ‡∏≠‡∏≠‡∏Å
u_split = u_strip.split(',') # ['x1','y1','z1'] # list(str)
v_split = v_strip.split(',') # ['x2','y2','z2'] # list(str)

# Step 4 : process : ‡πÄ‡∏õ‡∏•‡∏µ‡πà‡∏¢‡∏ô data type ‡∏Ç‡∏≠‡∏á‡∏™‡∏°‡∏≤‡∏ä‡∏¥‡∏Å‡∏†‡∏≤‡∏¢‡πÉ‡∏ô list ‡∏à‡∏≤‡∏Å str ‡πÉ‡∏´‡πâ‡πÄ‡∏õ‡πá‡∏ô int
x1 = float(u_split[0]) #str (‡∏ï‡∏±‡∏ß‡πÄ‡∏•‡∏Ç) -> float
y1 = float(u_split[1]) #str (‡∏ï‡∏±‡∏ß‡πÄ‡∏•‡∏Ç) -> float
z1 = float(u_split[2]) #str (‡∏ï‡∏±‡∏ß‡πÄ‡∏•‡∏Ç) -> float

x2 = float(v_split[0]) #str (‡∏ï‡∏±‡∏ß‡πÄ‡∏•‡∏Ç) -> float
y2 = float(v_split[1]) #str (‡∏ï‡∏±‡∏ß‡πÄ‡∏•‡∏Ç) -> float
z2 = float(v_split[2]) #str (‡∏ï‡∏±‡∏ß‡πÄ‡∏•‡∏Ç) -> float

# Step 5 : process : ‡∏ô‡∏≥ ‡∏≠‡∏õ‡∏Å. vector (x,y,z) ‡∏Ç‡∏≠‡∏á‡πÄ‡πÄ‡∏ï‡πà‡∏•‡∏∞‡πÄ‡πÄ‡∏Å‡∏ô ‡∏°‡∏≤‡∏£‡∏ß‡∏°‡∏Å‡∏±‡∏ô
x_total = x1 + x2 # float + float = float
y_total = y1 + y2 # float + float = float
z_total = z1 + z2 # float + float = float

# Step 6 : output : print ‡∏ï‡∏≤‡∏° format ‡πÇ‡∏à‡∏ó‡∏¢‡πå (‡∏™‡∏±‡∏á‡πÄ‡∏Å‡∏ï‡∏î‡∏µ‡πÜ ‡∏ó‡∏∏‡∏Å‡∏™‡∏°‡∏≤‡∏ä‡∏¥‡∏Å ‡∏à‡∏∞‡πÄ‡∏õ‡πá‡∏ô float)

# 6.1 : define ‡∏ï‡∏õ.‡∏Å‡πà‡∏≠‡∏ô ‡∏Å‡∏±‡∏ô‡πÇ‡∏Ñ‡πâ‡∏î‡∏¢‡∏≤‡∏ß‡πÄ‡∏Å‡∏¥‡∏ô

format_u = '[' + str(x1) + ', ' + str(y1) + ', ' + str(z1) + ']' # ‡∏≠‡∏¢‡πà‡∏≤‡∏•‡∏∑‡∏° print ‡∏ï‡πâ‡∏≠‡∏á data type ‡πÄ‡∏î‡∏µ‡∏¢‡∏ß‡∏Å‡∏±‡∏ô Ex print(str+int) ‡∏î‡πâ‡∏ß‡∏¢‡∏Å‡∏±‡∏ô‡πÑ‡∏°‡πà‡πÑ‡∏î‡πâ
format_v = '[' + str(x2) + ', ' + str(y2) + ', ' + str(z2) + ']'
format_total = '[' + str(x_total) + ', ' + str(y_total) + ', ' + str(z_total) + ']'

# 6.2 : print ‡∏ï‡∏≤‡∏° format
print(format_u , '+' , format_v , '=' , format_total) #str ‡∏£‡∏ß‡∏°‡∏Å‡∏±‡∏ô

## ‡πÇ‡∏à‡∏ó‡∏¢‡πå‡πÄ‡∏´‡∏•‡∏µ‡πà‡∏¢‡∏° : ‡∏î‡∏π‡∏î‡∏µ‡πÜ‡∏ß‡πà‡∏≤‡∏°‡∏±‡∏ô‡πÄ‡∏ß‡πâ‡∏ô‡∏ß‡∏£‡∏£‡∏Ñ‡∏ï‡∏£‡∏á ,  ‡∏î‡πâ‡∏ß‡∏¢ => ‡πÉ‡∏™‡πà‡πÄ‡∏õ‡πá‡∏ô ', ' ‡πÅ‡∏ó‡∏ô => ‡πÑ‡∏î‡πâ‡∏à‡∏≤‡∏Å 0 ‡πÄ‡∏õ‡πá‡∏ô 100
```

```python
# Key : ‡πÄ‡∏£‡∏≤‡∏à‡∏∞‡∏£‡∏±‡∏ö‡∏Ç‡πâ‡∏≠‡∏°‡∏π‡∏•‡∏¢‡∏±‡∏á‡πÑ‡∏á? ‡πÉ‡∏ô‡πÄ‡∏°‡∏∑‡πà‡∏≠‡∏õ‡πâ‡∏≠‡∏ô‡πÄ‡∏Ç‡πâ‡∏≤‡∏°‡∏≤‡πÉ‡∏ô‡∏£‡∏π‡∏õ‡πÄ‡πÄ‡∏ö‡∏ö '[x,y,z]' ‡πÉ‡∏´‡πâ‡πÄ‡∏õ‡πá‡∏ô list ‡πÄ‡∏û‡∏∑‡πà‡∏≠‡∏õ‡∏£‡∏∞‡∏°‡∏ß‡∏•‡∏ú‡∏•‡πÑ‡∏î‡πâ

# ‡∏ß‡∏¥‡∏ò‡∏µ‡∏ï‡∏£‡∏á : ‡πÄ‡∏£‡∏≤‡πÄ‡∏•‡∏¢‡∏à‡∏∞‡∏ó‡∏≥‡∏Å‡∏≤‡∏£ strip [ ‡πÄ‡πÄ‡∏•‡∏∞ ] ‡∏≠‡∏≠‡∏Å‡∏Å‡πà‡∏≠‡∏ô ‡πÄ‡πÄ‡∏•‡πâ‡∏ß split ‡∏î‡πâ‡∏ß‡∏¢ , (‡∏ó‡∏≥‡πÑ‡∏ß‡πâ‡∏î‡πâ‡∏≤‡∏ô‡∏ö‡∏ô‡πÄ‡πÄ‡∏•‡πâ‡∏ß)

## ‡∏ß‡∏¥‡∏ò‡∏µ Challenge : slice ‡πÄ‡∏≠‡∏≤‡∏ä‡πà‡∏ß‡∏á‡∏ó‡∏µ‡πà‡πÄ‡∏£‡∏≤‡∏™‡∏ß‡∏ô‡πÉ‡∏à ‡πÄ‡πÄ‡∏•‡πâ‡∏ß‡∏ó‡∏≥‡∏Å‡∏≤‡∏£ split ‡∏î‡πâ‡∏ß‡∏¢ ,

    # ‡∏ñ‡πâ‡∏≤‡∏™‡∏±‡∏á‡πÄ‡∏Å‡∏ï‡∏î‡∏π‡∏î‡∏µ‡πÜ : ‡∏à‡∏∞‡πÄ‡∏´‡πá‡∏ô‡∏ß‡πà‡∏≤ ‡πÄ‡∏£‡∏≤‡∏£‡∏±‡∏ö input ‡πÄ‡∏õ‡πá‡∏ô str ‡πÄ‡∏Ç‡πâ‡∏≤‡∏°‡∏≤ -> ‡πÄ‡∏£‡∏≤ index ‡πÄ‡∏≠‡∏≤‡πÄ‡πÄ‡∏Ñ‡πà str ‡∏ó‡∏µ‡πà‡πÄ‡∏£‡∏≤‡∏™‡∏ô‡πÉ‡∏à‡πÑ‡∏î‡πâ ‡πÄ‡πÄ‡∏•‡πâ‡∏ß‡πÄ‡∏£‡∏≤‡∏Å‡πá split ‡πÑ‡∏î‡πâ‡πÄ‡∏•‡∏¢
    
    ## ‡∏Ñ‡∏∑‡∏≠ : index ‡πÄ‡∏≠‡∏≤‡∏ä‡πà‡∏ß‡∏á‡∏ó‡∏µ‡πà‡πÑ‡∏°‡πà‡∏°‡∏µ‡∏ß‡∏á‡πÄ‡∏•‡πá‡∏ö‡∏ï‡∏¥‡∏î‡∏≠‡∏¢‡∏π‡πà  Ex : '[1,2,3,4]' --> ‡πÄ‡∏£‡∏≤‡∏à‡∏∞‡πÄ‡∏£‡∏¥‡πà‡∏°‡∏ô‡∏±‡∏ö index ‡∏†‡∏≤‡∏¢‡πÉ‡∏ï‡πâ str
    # index ‡∏ï‡∏±‡∏ß‡∏ó‡∏µ‡πà (‡∏ó‡∏≤‡∏á‡∏ö‡∏ß‡∏Å)                     012345678
    # index ‡∏ï‡∏±‡∏ß‡∏ó‡∏µ‡πà (‡∏ó‡∏≤‡∏á‡∏•‡∏ö)                   -  987654321
    
    ## ‡∏ó‡∏µ‡πà‡πÄ‡∏£‡∏≤‡∏™‡∏ô‡πÉ‡∏à ‡∏Ñ‡∏∑‡∏≠ '1,2,3,4' ‡πÄ‡∏£‡∏≤‡πÄ‡∏•‡∏ô‡∏¢ index ‡πÄ‡∏≠‡∏≤ ‡∏ï‡∏ï. index 1 ‡∏ñ‡∏∂‡∏á -1 (‡πÄ‡∏≠‡∏≤‡∏Å‡πà‡∏≠‡∏ô-1 = -2)
    # ‡πÇ‡∏Ñ‡πâ‡∏î‡∏à‡∏∞‡∏≠‡∏¢‡∏π‡πà‡πÉ‡∏ô‡∏£‡∏π‡∏õ u[1:-1] (‡πÑ‡∏°‡πà‡πÄ‡∏≠‡∏≤ ‡∏ï‡∏ô. -1 ‡∏à‡∏ö‡∏ó‡∏µ‡πà ‡∏ï‡∏ô. -2)
    
# 1 : ‡∏£‡∏±‡∏ö input ‡∏û‡∏£‡πâ‡∏≠‡∏° slice ‡∏ï‡∏ô. (‡πÑ‡∏°‡πà‡πÄ‡∏≠‡∏≤ () )‡πÄ‡πÄ‡∏•‡πâ‡∏ß‡∏Ñ‡πà‡∏≠‡∏¢ split ‡∏î‡πâ‡∏ß‡∏¢ ,
u = input('Vector U : ')[1:-1].split(',') # str -> slice -> list (of str) -> ['x1','y1','z1']
v = input('Vector V : ')[1:-1].split(',') # str -> slice -> list (of str) -> ['x1','y1','z1']

# 2 : ‡∏ó‡∏≥ operation ‡∏Ç‡∏≠‡∏á vector (‡∏™‡∏£‡πâ‡∏≤‡∏á list add_vec ‡πÄ‡∏õ‡πá‡∏ô‡∏ú‡∏•‡∏£‡∏ß‡∏°‡∏Ç‡∏≠‡∏á‡∏ó‡∏±‡πâ‡∏á 3 ‡πÄ‡πÄ‡∏Å‡∏ô‡∏Ç‡∏≠‡∏á u & v)

# 2.1 : ‡∏≠‡∏¢‡πà‡∏≤‡∏•‡∏∑‡∏°!! ‡πÄ‡πÄ‡∏õ‡∏•‡∏á‡∏Ñ‡πà‡∏≤ u ,v ‡πÉ‡∏´‡πâ‡πÄ‡∏õ‡πá‡∏ô list (of float) ‡∏Å‡πà‡∏≠‡∏ô
u = [float(u[0]) , float(u[1]) , float(u[2])] # ['x1','y1','z1'] -> [x1 ,y1 ,z1]
v = [float(v[0]) , float(v[1]) , float(v[2])] # ['x2','y2','z2'] -> [x2 ,y2 ,z2]

# 2.2 : ‡∏ó‡∏≥ op. ‡∏Ç‡∏≠‡∏á vector
add_vec = [u[0]+v[0] , u[1]+v[1] , u[2]+v[2]]
    # ‡∏Ñ‡∏∑‡∏≠  [x1 + x2   , y1 + y2   , z1 + z2 ] = [xt,yt,zt] (float) (# Note : float(' 4') = 4.0 (‡πÑ‡∏°‡πà‡∏™‡∏ô‡πÉ‡∏à‡∏ä‡πà‡∏≠‡∏á‡∏ß‡πà‡∏≤‡∏á))

# 3 : print output
print(u , '+' , v , '=' , add_vec) # u,v = str & add_vec = list (of float)
```

### Task 02_08 : Demical to Fractions (3 Star)


```python
# 1 : import math libarry for gcd function
import math
# 2 : input info as str
integer , notrep , rep = input().split(',') # str

# 3 : find numerator & denominator
    # numerator = before_rep (rep concat notrep) - notrep
    # denominator = 99..(length of rep) 00....(length of not_rep)
    
numerator = ( int((notrep or '') + rep) ) - int(notrep or '0')

denominator = int( '9' * len(rep) + '0' * len(notrep) )

# 4 : add integer to fractions
numerator += (int(integer) * denominator)

# 5 : find gcd of number
divider = math.gcd(numerator , denominator)

# 6 : print out string
print(str(numerator//divider) + ' / ' + str(denominator//divider) )
```

```python
# Sol1 : ‡πÉ‡∏ä‡πâ‡∏ß‡∏¥‡∏ò‡∏µ‡∏ï‡∏£‡∏á
def check_digit(n) : # Step 1 : ‡πÄ‡∏£‡∏≤‡∏à‡∏∞‡∏£‡∏±‡∏ö‡∏Ñ‡πà‡∏≤ input ‡∏Ñ‡∏∑‡∏≠ n ‡∏ã‡∏∂‡πà‡∏á‡∏Ñ‡∏∑‡∏≠‡πÄ‡∏•‡∏Ç 12 ‡∏´‡∏•‡∏±‡∏Å (parameter)
    
    # Step 2 : ‡πÄ‡∏£‡∏≤‡∏à‡∏∞ index ‡πÄ‡πÄ‡∏ï‡πà‡∏•‡∏∞‡∏ï‡∏±‡∏ß -> ‡πÄ‡πÄ‡∏õ‡∏•‡∏á‡πÉ‡∏´‡πâ‡πÄ‡∏õ‡πá‡∏ô int -> ‡πÉ‡∏™‡πà‡∏™‡∏π‡∏ï‡∏£ ‡∏´‡∏≤ ‡∏´‡∏•‡∏±‡∏Å‡∏™‡∏∏‡∏î‡∏ó‡πâ‡∏≤‡∏¢
    lastdigit = (11- (13*int(n[0]) + 12*int(n[1]) + 11*int(n[2]) + 10*int(n[3]) + 9*int(n[4]) + 8*int(n[5]) + 7*int(n[6]) + 6*int(n[7]) + 5*int(n[8]) + 4*int(n[9]) + 3*int(n[10]) + 2*int(n[11])) % 11 ) % 10
    
    # Step 3 : return ‡∏Ñ‡πà‡∏≤ ‡∏´‡∏•‡∏±‡∏Å‡∏™‡∏∏‡∏î‡∏ó‡πâ‡∏≤‡∏¢
    return lastdigit

exec(input()) # ‡πÉ‡∏™‡πà fn ‡∏ô‡∏µ‡πâ ‡πÄ‡∏û‡∏∑‡πà‡∏≠‡∏ó‡∏≥‡∏Å‡∏≤‡∏£ print input ‡πÄ‡∏û‡∏∑‡πà‡∏≠‡πÄ‡∏ä‡πá‡∏Ñ fn
```

### Task 02_11 : Add Vector (Function) (2 Star)


```python
def add_vector (u,v) : # 1 : ‡∏£‡∏±‡∏ö parameter u,v ‡πÄ‡∏õ‡πá‡∏ô list ‡∏ó‡∏µ‡πà‡∏°‡∏µ‡∏™‡∏°‡∏≤‡∏ä‡∏¥‡∏Å 3 ‡∏ï‡∏±‡∏ß Ex : u = [x1,y1,z1] , y = [v1,v2,v3] 
    
    # 2 : ‡∏™‡∏£‡πâ‡∏≤‡∏á ‡∏ï‡∏õ.‡πÄ‡∏û‡∏∑‡πà‡∏≠‡πÄ‡πÄ‡∏ó‡∏ô‡πÄ‡πÄ‡∏ï‡πà‡∏•‡∏∞ index => ‡πÄ‡πÄ‡∏õ‡∏•‡∏á‡∏Ñ‡πà‡∏≤‡πÄ‡∏õ‡πá‡∏ô float
    u1 , u2 , u3 = float(u[0]) , float(u[1]) , float(u[2])
    v1 , v2 , v3 = float(v[0]) , float(v[1]) , float(v[2])
    
    # 3 : ‡∏£‡∏ß‡∏° vector
    add_result = [u1+v1 , u2+v2 , u3+v3]
    
    # 4 : return ‡∏Ñ‡πà‡∏≤‡∏à‡∏≤‡∏Å‡∏Å‡∏≤‡∏£‡∏£‡∏ß‡∏° vector
    return add_result

# 5 : execute input ‡πÄ‡∏û‡∏∑‡πà‡∏≠‡∏ó‡∏≥‡∏Å‡∏≤‡∏£ print input ‡πÄ‡∏û‡∏∑‡πà‡∏≠‡πÄ‡∏ä‡πá‡∏Ñ fn
exec(input())    
```
