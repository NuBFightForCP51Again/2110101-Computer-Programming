### Task 01_01 : Stirling Factorial (1 Star)


```python
# Logic : ‡∏£‡∏±‡∏ö‡∏Ñ‡πà‡∏≤ n! -> ‡πÄ‡∏Ç‡πâ‡∏≤‡∏™‡∏π‡∏ï‡∏£ ‡∏Ç‡∏≠‡∏ö‡∏ö‡∏ô-‡∏•‡πà‡∏≤‡∏á ! -> print ‡∏Ñ‡πà‡∏≤‡∏Ç‡∏≠‡∏ö‡∏≠‡∏≠‡∏Å‡∏°‡∏≤
# Step : 
# 1. ‡∏£‡∏±‡∏ö input n (int)
# 2. ‡∏Å‡∏≥‡∏´‡∏ô‡∏î ‡∏ï‡∏õ. n_min , n_max (‡πÄ‡∏Ç‡πâ‡∏≤‡∏™‡∏π‡∏ï‡∏£)
# 3. print ‡∏Ç‡∏≠‡∏ö‡∏ö‡∏ô , ‡∏Ç‡∏≠‡∏ö‡∏•‡πà‡∏≤‡∏á

# Input : ‡∏£‡∏±‡∏ö input n (int) + import math
import math # ‡πÄ‡∏û‡∏∑‡πà‡∏≠‡πÉ‡∏ä‡πâ‡∏ü‡∏±‡∏á‡∏Å‡πå‡∏ä‡∏±‡πà‡∏ô‡∏ó‡∏≤‡∏á‡∏Ñ‡∏ì‡∏¥‡∏ï‡∏®‡∏≤‡∏™‡∏ï‡∏£‡πå
n = int(input()) #number for n! (int)

# Process : ‡∏Å‡∏≥‡∏ô‡∏´‡∏î‡∏Ñ‡πà‡∏≤ n_min , n_max
n_floor = math.sqrt(2*math.pi) * n**(n+0.5) * math.exp(-n + 1/((12*n)+1))
n_ceil = math.sqrt(2*math.pi) * n**(n+0.5) * math.exp(-n + 1/(12*n))

# Output : print ‡∏Ç‡∏≠‡∏ö‡∏ö‡∏ô , ‡∏Ç‡∏≠‡∏ö‡∏•‡πà‡∏≤‡∏á ‡∏Ç‡∏≠‡∏á n!
print(n_floor) #float
print(n_ceil) #float

## ‡∏≠‡∏¢‡πà‡∏≤‡∏•‡∏∑‡∏°!! ‡πÄ‡∏ß‡∏•‡∏≤‡πÄ‡∏≠‡∏≤‡πÑ‡∏õ‡∏•‡∏á‡πÉ‡∏ô Grader -> ‡∏•‡∏ö command ‡πÉ‡∏ô input() ‡∏≠‡∏≠‡∏Å‡∏î‡πâ‡∏ß‡∏¢

# Test Case : 
# case 1 : n = 1 -> 1! -> n_floor = 0.995870161462797 , n_ceil = 1.0022744491822266
# case 2 : n = 5 -> 5! -> n_floor = 119.96985395920886 , n_ceil = 120.00263708619694
# case 3 : n = 50 -> 50! -> n_floor = 3.041400953459947e+64 , n_ceil = 3.041409387750485e+64
# case 4 : n = 100 -> 100! -> n_floor = 9.332615094728949e+157 , n_ceil = 9.332621570317617e+157
```

### Solution 01_01 : Stirling Factorial (1 Star)

```python
# 1 : import math + ‡∏£‡∏±‡∏ö‡∏Ñ‡πà‡∏≤ n (int) ‡πÄ‡∏û‡∏∑‡πà‡∏≠‡∏´‡∏≤ n!
import math
n = int(input('Enter number for n! : '))

# 2 : ‡πÄ‡∏Ç‡πâ‡∏≤‡∏™‡∏π‡∏ï‡∏£ ‡πÄ‡∏û‡∏∑‡πà‡∏≠‡∏´‡∏≤‡∏Ñ‡πà‡∏≤ lower & upper ‡∏Ç‡∏≠‡∏á n!

lower = math.sqrt(2*math.pi) * (n ** (n + 0.5)) \
    * math.e**(-n + (1/ (12*n + 1) )) # \ ‡∏Ñ‡∏∑‡∏≠‡∏Ç‡∏∂‡πâ‡∏ô‡∏ö‡∏£‡∏£‡∏ó‡∏±‡∏î‡πÉ‡∏´‡∏°‡πà
            
upper = math.sqrt(2*math.pi) * (n ** (n + 0.5)) \
    * math.e**(-n + (1/ (12*n) )) # \ ‡∏Ñ‡∏∑‡∏≠‡∏Ç‡∏∂‡πâ‡∏ô‡∏ö‡∏£‡∏£‡∏ó‡∏±‡∏î‡πÉ‡∏´‡∏°‡πà
    
# 3 : print ‡∏Ñ‡πà‡∏≤‡∏≠‡∏≠‡∏Å‡∏°‡∏≤
print('Lower of n! is : ' , lower)
print('Upper of n! is : ' , upper)

## ‡∏≠‡∏¢‡πà‡∏≤‡∏•‡∏∑‡∏°‡∏•‡∏ö‡∏Ñ‡∏≥‡∏û‡∏π‡∏î‡∏≠‡∏≠‡∏Å‡πÉ‡∏ô grader
```

### Task 01_02 : Quadratic Root (1 Star)

### Solution 01_02 : Quadratic Roots (1 Star)

```python
# Logic : ‡πÄ‡πÄ‡∏Å‡πâ‡∏™‡∏°‡∏Å‡∏≤‡∏£‡∏Å‡∏≥‡∏•‡∏±‡∏á 2

# 1 : ‡∏£‡∏±‡∏ö‡∏Ñ‡πà‡∏≤ a,b,c (float) + import math
import math
a = float(input('Input A for x^2 degree : ')) #float
b = float(input('Input B for x^1 degree : ')) #float
c = float(input('Input C for x^0 degree : ')) #float

# 2 : ‡πÄ‡∏Ç‡πâ‡∏≤‡∏™‡∏π‡∏ï‡∏£ ‡πÄ‡πÄ‡∏Å‡πâ‡∏™‡∏°‡∏Å‡∏≤‡∏£
discriminant = (b*b - 4*a*c) 

## ‡∏£‡∏∞‡∏ß‡∏±‡∏á ! ‡∏≠‡∏¢‡πà‡∏≤‡∏•‡∏∑‡∏°‡∏î‡∏π‡∏î‡∏µ‡πÜ ‡∏ß‡πà‡∏≤‡πÄ‡∏£‡∏≤ input data ‡∏õ‡∏£‡∏∞‡πÄ‡∏†‡∏ó‡πÑ‡∏´‡∏ô‡πÄ‡∏Ç‡πâ‡∏≤‡∏°‡∏≤ (str*str ‡∏ó‡∏≥ math op. ‡πÑ‡∏°‡πà‡πÑ‡∏î‡πâ)

x1 = ( -b + math.sqrt(discriminant) ) / (2*a)
x2 = ( -b - math.sqrt(discriminant) ) / (2*a)

ans1 = round(x1 , 3)
ans2 = round(x2 , 3)

# 3 : print x1 , x2
print('Answer 1 : ' , ans1 , 'Answer 2 : , ' , ans2) #print ‡∏ö‡∏ô‡∏ö‡∏£‡∏£‡∏ó‡∏±‡∏î‡πÄ‡∏î‡∏µ‡∏¢‡∏ß‡∏Å‡∏±‡∏ô

## ‡∏à‡∏£‡∏¥‡∏á‡πÜ‡πÄ‡∏£‡∏≤‡∏™‡∏≤‡∏°‡∏≤‡∏£‡∏ñ‡∏¢‡∏∏‡∏ö ‡∏Å‡∏≤‡∏£ round ‡πÉ‡∏ô x1 , x2 ‡πÑ‡∏î‡πâ‡πÄ‡∏•‡∏¢
# Ex : 
# Ex : 

## ‡∏≠‡∏¢‡πà‡∏≤‡∏•‡∏∑‡∏° ! ‡∏ù‡∏∂‡∏Å‡∏Å‡∏≤‡∏£‡∏Ñ‡∏¥‡∏î testcases ‡πÑ‡∏ß‡πâ‡∏ß‡πà‡∏≤‡∏™‡∏≤‡∏°‡∏≤‡∏£‡∏ñ‡πÄ‡∏Å‡∏¥‡∏î‡∏≠‡∏∞‡πÑ‡∏£‡∏Ç‡∏∂‡πâ‡∏ô‡πÑ‡∏î‡πâ‡∏ö‡πâ‡∏≤‡∏á ‡πÄ‡πÄ‡∏•‡πâ‡∏ß‡∏î‡∏π‡∏ß‡πà‡∏≤‡πÄ‡∏£‡∏≤‡∏ú‡∏¥‡∏î‡∏ï‡∏£‡∏á‡πÑ‡∏´‡∏ô
```

### Solution : 01_03 Strange Expression (2 Star)

```python
# Key : ‡πÄ‡πÄ‡∏¢‡∏Å‡∏Ñ‡∏¥‡∏î‡∏ó‡∏µ‡∏•‡∏∞‡∏™‡πà‡∏ß‡∏ô ‡∏Å‡∏±‡∏ô‡∏™‡∏±‡∏ö‡∏™‡∏ô

# 1 : import math 
import math

# 2 : ‡πÄ‡πÄ‡∏¢‡∏Å‡∏Ñ‡∏¥‡∏î‡∏ó‡∏µ‡∏•‡∏∞‡∏™‡πà‡∏ß‡∏ô
a = math.factorial(10) / (8 ** 8)
b = math.log(9.7 , math.e) # ‡∏õ‡∏Å‡∏ï‡∏¥‡πÄ‡∏õ‡πá‡∏ô‡∏ê‡∏≤‡∏ô e ‡πÑ‡∏°‡πà‡∏ï‡πâ‡∏≠‡∏á‡πÉ‡∏™‡πà‡∏ê‡∏≤‡∏ô‡∏Å‡πá‡πÑ‡∏î‡πâ (cmd : math.log(x,base)) ## ‡πÉ‡∏ä‡πâ‡πÄ‡∏õ‡πá‡∏ô math.e naja
c = (7 / math.sqrt(71)) - math.sin(40/180 * math.pi) # ‡∏´‡∏£‡∏∑‡∏≠‡∏à‡∏∞‡πÉ‡∏ä‡πâ math.radians(40) ‡∏Å‡πá‡πÑ‡∏î‡πâ
d = 1.2 ** ((2.3) ** (1/3))

##  ‡∏£‡∏∞‡∏ß‡∏±‡∏á‡∏°‡∏∏‡∏° 40 ‡∏≠‡∏á‡∏®‡∏≤ => math.sin() ‡πÉ‡∏ä‡πâ‡∏´‡∏ô‡πà‡∏ß‡∏¢ rad -> ‡πÄ‡πÄ‡∏õ‡∏•‡∏á‡πÄ‡∏õ‡πá‡∏ô rad ‡∏Å‡πà‡∏≠‡∏ô ‡πÄ‡πÄ‡∏•‡πâ‡∏ß‡∏Ñ‡πà‡∏≠‡∏¢‡πÉ‡∏ä‡πâ

# 3 : ‡∏£‡∏ß‡∏°‡∏£‡πà‡∏≤‡∏á
ans = (math.pi - a + (b ** c) )/d

# 4 : print ans
print( round(ans , 6) )
```

### Task 01_05 : Duration (3 Star)


```python
# Logic : ‡∏´‡∏≤ ‡∏ä‡πà‡∏ß‡∏á‡πÄ‡∏ß‡∏•‡∏≤‡∏ó‡∏µ‡πà‡∏ï‡πà‡∏≤‡∏á‡∏Å‡∏±‡∏ô (‡πÄ‡πÄ‡∏õ‡∏•‡∏á‡πÄ‡∏õ‡πá‡∏ô sec ‡∏Å‡πà‡∏≠‡∏ô ‡πÄ‡πÄ‡∏•‡πâ‡∏ß‡πÄ‡πÄ‡∏õ‡∏•‡∏á‡πÄ‡∏õ‡πá‡∏ô  ‡πâ‡∏û)
# Step : 
# 1. ‡∏£‡∏±‡∏ö‡∏Ñ‡πà‡∏≤ h1,m1,s1 ‡πÄ‡πÄ‡∏•‡∏∞ h2,m2,s2 (int)  
# 2. ‡πÄ‡πÄ‡∏õ‡∏•‡∏á‡∏Ñ‡πà‡∏≤ h,m,s ‡πÉ‡∏´‡πâ‡πÄ‡∏õ‡πá‡∏ô t1 , t2 ‡∏ï‡∏≤‡∏°‡∏•‡∏≥‡∏î‡∏±‡∏ö
# 3. define ‡∏ï‡∏õ. dt (difference time) = t2 - t1
# 4. ‡πÄ‡πÄ‡∏õ‡∏•‡∏á dt ‡πÉ‡∏´‡πâ‡πÄ‡∏õ‡πá‡∏ô dh,dm,ds (‡∏´‡∏≤‡∏£‡∏õ‡∏±‡∏î‡πÄ‡∏®‡∏©)
# 5. ‡πÄ‡πÄ‡∏õ‡∏•‡∏á‡∏Ñ‡πà‡∏≤ dh ‡∏à‡∏≤‡∏Å - ‡πÉ‡∏´‡πâ‡πÄ‡∏õ‡πá‡∏ô + => (24 - dh)

# Step 1 : ‡∏£‡∏±‡∏ö input h1,m1,s1,h2,m2,s2 (int)
h1 = int(input()) 
m1 = int(input())
s1 = int(input())

h2 = int(input())
m2 = int(input())
s2 = int(input())

# Step 2 : ‡πÄ‡πÄ‡∏õ‡∏•‡∏á‡∏Ñ‡πà‡∏≤ h,m,s ‡πÄ‡∏õ‡πá‡∏ô t 
t1 = h1*60*60 + m1*60 + s1 #time total (sec) initial
t2 = h2*60*60 + m2*60 + s2 #time total (sec) final

# Step 3 : define dt (t_final - t_initial) & check ‡πÄ‡∏á‡∏∑‡πà‡∏≠‡∏ô‡πÑ‡∏Ç t2 ‡∏Å‡∏±‡∏ö t1
dt = t2 - t1

if t2 < t1 :
    dt = 86400 + (t2 - t1) #‡∏ö‡∏ß‡∏Å‡πÑ‡∏õ 1 ‡∏ß‡∏±‡∏ô ‡πÄ‡∏û‡∏∑‡πà‡∏≠‡πÄ‡∏Å‡∏¥‡∏î‡πÄ‡∏®‡∏©‡∏ß‡∏±‡∏ô

# Step 4 : ‡πÅ‡∏õ‡∏•‡∏á dt ‡πÉ‡∏´‡πâ‡πÄ‡∏õ‡πá‡∏ô dh , dm , ds
dh = dt // (60*60) #‡∏´‡∏≤‡∏£‡∏õ‡∏±‡∏î‡πÄ‡∏®‡∏© 3600 sec -> ‡πÄ‡∏≠‡∏≤ ‡∏à‡∏ô.hr
dt = dt - (dh * 3600) #‡∏•‡∏ö ‡∏à‡∏ô.‡∏ä‡∏±‡πà‡∏ß‡πÇ‡∏°‡∏á‡∏≠‡∏≠‡∏Å ‡πÄ‡∏´‡∏•‡∏∑‡∏≠‡πÄ‡∏®‡∏© hr (sec) ## ‡∏ï‡∏≠‡∏ô‡∏ô‡∏µ‡πâ dt ‡πÄ‡∏´‡∏•‡∏∑‡∏≠‡πÄ‡∏®‡∏© hr ‡πÄ‡πÄ‡∏•‡πâ‡∏ß -> ‡πÄ‡πÄ‡∏õ‡∏•‡∏á‡∏ï‡πà‡∏≠‡πÄ‡∏õ‡πá‡∏ô m,s

dm = dt // 60 #‡∏´‡∏≤‡∏£‡∏õ‡∏±‡∏î‡πÄ‡∏®‡∏© 60 sec -> ‡πÄ‡∏≠‡∏≤ ‡∏à‡∏ô. min
dt = dt - (dm * 60) #‡∏•‡∏ö ‡∏à‡∏ô.‡∏ô‡∏≤‡∏ó‡∏µ‡∏≠‡∏≠‡∏Å ‡πÄ‡∏´‡∏•‡∏∑‡∏≠‡πÄ‡∏®‡∏© min (sec) ## ‡∏ï‡∏≠‡∏ô‡∏ô‡∏µ‡πâ dt ‡πÄ‡∏´‡∏•‡∏∑‡∏≠‡πÄ‡∏®‡∏© min ‡πÅ‡∏•‡πâ‡∏ß-> ‡πÑ‡∏°‡πà‡∏ï‡πâ‡∏≠‡∏á‡πÄ‡πÄ‡∏õ‡∏•‡∏á‡∏ï‡πà‡∏≠‡πÄ‡πÄ‡∏•‡πâ‡∏ß ‡πÄ‡∏û‡∏£‡∏≤‡∏∞‡∏ó‡∏µ‡πà‡πÄ‡∏´‡∏•‡∏∑‡∏≠‡πÄ‡∏õ‡πá‡∏ô‡∏´‡∏ô‡πà‡∏ß‡∏¢ sec

ds = dt #‡πÄ‡∏®‡∏© min = ‡∏à‡∏ô.sec ‡∏ó‡∏µ‡πà‡πÄ‡∏´‡∏•‡∏∑‡∏≠ ‡∏Å‡∏≥‡∏´‡∏ô‡∏î‡πÉ‡∏´‡πâ‡πÄ‡∏õ‡πá‡∏ô ds ‡πÑ‡∏î‡πâ‡πÄ‡∏•‡∏¢

# Step 5 : print ‡∏ï‡∏≤‡∏° format ‡πÇ‡∏à‡∏ó‡∏¢‡πå
print( str(dh) + ':' + str(dm) + ':' + str(ds) ) #‡πÄ‡∏Ç‡∏µ‡∏¢‡∏ô‡πÉ‡∏ô format HH : MM : SS


## ‡∏•‡∏≠‡∏á‡πÄ‡∏≠‡∏≤‡πÑ‡∏õ run ‡πÉ‡∏ô grader ‡πÄ‡πÄ‡∏•‡πâ‡∏ß

# Note : Sol 1 : ‡∏•‡∏≠‡∏á‡πÉ‡∏ä‡πâ dt = abs(t2 - t1) ‡πÑ‡∏î‡πâ‡πÄ‡πÄ‡∏Ñ‡πà 40 pts ‡πÄ‡πÄ‡∏™‡∏î‡∏á‡∏ß‡πà‡∏≤ ‡∏¢‡∏±‡∏á‡∏°‡∏µ‡∏õ‡∏±‡∏ç‡∏´‡∏≤‡∏≠‡∏¢‡∏π‡πà‡πÉ‡∏ô‡∏ö‡∏≤‡∏á testcases -> ‡∏•‡∏≠‡∏á‡πÉ‡∏ä‡πâ mod ‡∏î‡∏π (‡∏ô‡πâ‡∏≥‡πÄ‡∏≠‡∏Å‡πÄ‡πÄ‡∏ô‡∏∞‡∏ô‡∏≥)
#        Sol 2 : ‡πÄ‡∏à‡∏≠‡∏õ‡∏±‡∏ç‡∏´‡∏≤‡∏ß‡πà‡∏≤ dt ‡πÄ‡∏õ‡πá‡∏ô‡∏•‡∏ö‡∏≠‡∏¢‡∏π‡πà -> ‡πÄ‡πÄ‡∏Å‡πâ‡πÇ‡∏î‡∏¢ + 24 ‡πÄ‡∏Ç‡πâ‡∏≤‡πÑ‡∏õ -> dh = 24 + (h2-h1) -> ‡πÑ‡∏î‡πâ 50 ‡∏Ñ‡∏∞‡πÄ‡πÄ‡∏ô‡∏ô -> ‡∏¢‡∏±‡∏á‡∏°‡∏µ‡∏õ‡∏±‡∏ç‡∏´‡∏≤‡∏≠‡∏¢‡∏π‡πà
#        Sol 3 : ‡πÄ‡∏à‡∏≠‡∏õ‡∏±‡∏ç‡∏´‡∏≤‡∏ß‡πà‡∏≤ dt ‡πÄ‡∏Å‡∏¥‡∏ô 25 ‡∏ä‡∏°.‡πÑ‡∏î‡πâ -> ‡πÄ‡πÄ‡∏Å‡πâ‡πÇ‡∏î‡∏¢ mod 24 ‡πÄ‡∏û‡∏∑‡πà‡∏≠‡πÉ‡∏´‡πâ‡πÑ‡∏°‡πà‡πÄ‡∏Å‡∏¥‡∏ô 24 -> ‡πÑ‡∏î‡πâ 70 ‡∏Ñ‡∏∞‡πÄ‡πÄ‡∏ô‡∏ô -> ‡πÄ‡∏î‡∏∞‡πÄ‡∏ä‡πá‡∏Ñ‡∏ö‡∏≤‡∏á testcases ‡∏ó‡∏µ‡πà‡∏ú‡∏¥‡∏î 
#        Sol 4 : ‡∏ï‡∏≠‡∏ô‡∏ô‡∏µ‡πâ‡πÄ‡∏à‡∏≠‡∏õ‡∏±‡∏ç‡∏´‡∏≤‡∏ß‡πà‡∏≤ dh ‡πÄ‡∏Å‡∏¥‡∏ô‡∏°‡∏≤ 1 ‡∏ï‡∏•‡∏≠‡∏î (‡∏ñ‡πâ‡∏≤ 23 -> ‡πÇ‡∏ä‡∏ß‡πå 0 , ‡∏ñ‡πâ‡∏≤ 1 -> ‡πÇ‡∏ä‡∏ß‡πå 2)

#        Sol 5 : ‡∏ñ‡πâ‡∏≤‡∏´‡∏≤‡∏Å t2 < t1 -> ‡πÄ‡∏ß‡∏•‡∏≤‡∏à‡∏∞‡∏ï‡∏¥‡∏î‡∏•‡∏ö -> ‡πÄ‡∏£‡∏≤‡∏à‡∏∞‡∏ö‡∏ß‡∏Å‡πÄ‡∏û‡∏¥‡πà‡∏°‡πÑ‡∏õ 1 ‡∏ß‡∏±‡∏ô ‡πÄ‡∏û‡∏∑‡πà‡∏≠‡∏ó‡∏≥‡πÉ‡∏´‡πâ‡πÄ‡∏Å‡∏¥‡∏î‡πÄ‡∏®‡∏©‡∏ß‡∏±‡∏ô -> ‡πÄ‡πÄ‡∏•‡πâ‡∏ß‡πÄ‡πÄ‡∏õ‡∏•‡∏á‡πÄ‡∏õ‡πá‡∏ô h,m,s => 100 ja
```

### Solution : 01_05 Duration (3 Star) By Dr.Somchai

```python
## Solution By SPJ
## ‡πÄ‡∏à‡∏≠‡∏õ‡∏±‡∏ç‡∏´‡∏≤‡∏ß‡πà‡∏≤ dh = ‡∏ú‡∏•‡∏ï‡πà‡∏≤‡∏á‡∏à‡∏ô.‡∏ä‡∏°. ‡πÄ‡∏õ‡πá‡∏ô‡∏•‡∏ö => ‡πÄ‡πÄ‡∏Å‡πâ‡πÇ‡∏î‡∏¢ +24 ‡πÄ‡πÄ‡∏•‡πâ‡∏ß mod 24 ‡πÄ‡∏û‡∏∑‡πà‡∏≠‡πÉ‡∏´‡πâ‡πÄ‡∏õ‡πá‡∏ô + (‡πÄ‡πÄ‡∏ï‡πà‡∏°‡∏±‡∏ô‡πÉ‡∏ä‡πâ‡πÑ‡∏î‡πâ‡πÄ‡πÄ‡∏Ñ‡πà‡∏ä‡∏°.)
## ‡πÄ‡∏£‡∏≤‡∏ï‡πâ‡∏≠‡∏á‡∏Å‡∏≤‡∏£‡∏ß‡∏¥‡∏ô‡∏≤‡∏ó‡∏µ‡∏≠‡∏≠‡∏Å‡∏°‡∏≤ -> ‡πÉ‡∏ä‡πâ‡∏ß‡∏¥‡∏ò‡∏µ‡πÄ‡∏î‡∏µ‡∏¢‡∏ß‡∏Å‡∏±‡∏ô‡∏Å‡∏±‡∏ö dh ‡πÄ‡πÄ‡∏ï‡πà‡∏à‡∏∞‡∏™‡∏£‡πâ‡∏≤‡∏á‡∏ï‡∏õ.‡∏à‡∏ô.‡∏ß‡∏¥‡∏ô‡∏≤‡∏ó‡∏µ‡πÉ‡∏ô 1 ‡∏ß‡∏±‡∏ô (86,400)‡∏Ç‡∏∂‡πâ‡∏ô‡∏°‡∏≤

# 1 : ‡∏£‡∏±‡∏ö input h,m,s ‡πÄ‡∏Ç‡πâ‡∏≤‡∏°‡∏≤
h1 = int(input())
m1 = int(input())
s1 = int(input())
h2 = int(input())
m2 = int(input())
s2 = int(input())

# 2 : ‡∏´‡∏≤‡∏ú‡∏•‡∏ï‡πà‡∏≤‡∏á‡πÄ‡∏ß‡∏•‡∏≤ + ‡∏ï‡∏±‡πâ‡∏á ‡∏ï‡∏õ. sec_in_day ‡∏Ç‡∏∂‡πâ‡∏ô‡∏°‡∏≤
# 2.1 : ‡πÄ‡πÄ‡∏õ‡∏•‡∏á‡πÄ‡∏õ‡πá‡∏ô‡∏ß‡∏¥‡∏ô‡∏≤‡∏ó‡∏µ‡πÉ‡∏´‡πâ‡∏´‡∏°‡∏î‡∏Å‡πà‡∏≠‡∏ô
t1 = 3600*h1 + 60*m1 + s1 #sec
t2 = 3600*h2 + 60*m2 + s2 #sec
dt = t2 - t1

# 2.2 : ‡πÄ‡πÄ‡∏õ‡∏•‡∏á‡∏™‡∏π‡∏ï‡∏£ ‡πÇ‡∏î‡∏¢‡∏ó‡∏µ‡πà‡∏ú‡∏•‡∏ï‡πà‡∏≤‡∏á‡πÄ‡∏õ‡πá‡∏ô + & ‡πÑ‡∏°‡πà‡πÄ‡∏Å‡∏¥‡∏î‡πÄ‡∏Ñ‡∏™‡πÄ‡∏®‡∏©‡πÄ‡∏Å‡∏¥‡∏ô
    # ‡πÄ‡πÄ‡∏õ‡∏•‡∏á‡∏£‡∏≠‡∏ö‡πÄ‡πÄ‡∏£‡∏Å : +86,400 ‡πÉ‡∏´‡πâ‡πÄ‡∏õ‡πá‡∏ô‡∏Ñ‡πà‡∏≤ + (‡∏ñ‡πâ‡∏≤‡πÄ‡∏Å‡∏¥‡∏ô ‡∏Å‡πá mod 86400 ‡∏à‡∏∞‡πÄ‡∏´‡∏•‡∏∑‡∏≠ dt)
dt = (86400 + t2 - t1)
    # ‡πÅ‡∏õ‡∏•‡∏á‡∏£‡∏≠‡∏ö 2 : mod 86400 -> ‡πÑ‡∏î‡πâ dt
dt = dt % 86400 

# 3 : ‡πÄ‡πÄ‡∏õ‡∏•‡∏á dt ‡πÉ‡∏´‡πâ‡∏≠‡∏¢‡∏π‡πà‡πÉ‡∏ô‡∏£‡∏π‡∏õ h,m,s
dh = dt // 3600 # ‡∏´‡∏≤‡∏£‡∏õ‡∏±‡∏î‡πÄ‡∏®‡∏© ‡πÄ‡∏≠‡∏≤ hr
dt = dt - (3600*dh)

dm = dt // 60 # ‡∏´‡∏≤‡∏£‡∏õ‡∏±‡∏î‡πÄ‡∏®‡∏© ‡πÄ‡∏≠‡∏≤ min
dt = dt - (60*dm)

ds = dt

# 4 : print ‡∏ï‡∏≤‡∏° format
print(str(dh) + ':' + str(dm) + ':' + str(ds)) 
```

### Task 01_07 : Cube Root (Function) (2 Star)



```python
# 1 : ‡∏™‡∏£‡πâ‡∏≤‡∏á fn sqrt_n (parameter x,n)
def sqrt_n_times(x,n) : # ‡∏Ñ‡∏∑‡∏ô‡∏Ñ‡πà‡∏≤ ‚àö
    ## Logic : ‡∏´‡∏≤‡∏£‡∏π‡∏õ‡∏ó‡∏±‡πà‡∏ß‡πÑ‡∏õ ‡∏Ç‡∏≠‡∏á ‚àö‚àö‚àö...x = x^(1/2^n) ; n = ‡∏à‡∏ô.sqrt
    return x ** (1/(2**n))
   
# 2 : ‡∏™‡∏£‡πâ‡∏≤‡∏á fn cube root (parameter y) 
def cube_root (y) :
    # ‡∏Ñ‡∏∑‡∏ô‡∏Ñ‡πà‡∏≤ y ** ((1/2^2)(1 + 1/2^2)(1 + 1/2^4)(1 + 1/2^8)(1 + 1/2^16)(1 + 1/2^32))
    # Logic : ‡∏™‡∏±‡∏á‡πÄ‡∏Å‡∏ï‡∏ß‡πà‡∏≤ ‡∏¢‡∏Å‡∏Å‡∏≥‡∏•‡∏±‡∏á‡∏ã‡πâ‡∏≠‡∏ô‡πÑ‡∏õ‡πÄ‡∏£‡∏∑‡πà‡∏≠‡∏¢‡πÜ => ‡πÄ‡πÄ‡∏•‡πâ‡∏ß‡∏¢‡∏±‡∏î fn ‡∏•‡∏á‡πÑ‡∏õ
    
    y1 = sqrt_n_times(y,2) # y ** (1/2^2)
    y2 = y1 * sqrt_n_times(y1,2)    # ‡πÄ‡∏≠‡∏≤ y1 ‡∏°‡∏≤ ** (1 + 1/2^2) = y1 * (y1**(1/2^2)) = y1 * fn sqrt
    y3 = y2 * sqrt_n_times(y2,4)    # ‡πÄ‡∏≠‡∏≤ y2 ‡∏°‡∏≤ ** (1 + 1/2^4) = y2 * (y2**(1/2^4)) = y2 * fn sqrt
    y4 = y3 * sqrt_n_times(y3,8)    # ‡πÄ‡∏≠‡∏≤ y3 ‡∏°‡∏≤ ** (1 + 1/2^8) = y3 * (y3**(1/2^2)) = y3 * fn sqrt
    y5 = y4 * sqrt_n_times(y4,16)   # ‡πÄ‡∏≠‡∏≤ y4 ‡∏°‡∏≤ ** (1 + 1/2^16) = y4 * (y4**(1/2^2)) = y4 * fn sqrt
    y6 = y5 * sqrt_n_times(y5,32)   # ‡πÄ‡∏≠‡∏≤ y5 ‡∏°‡∏≤ ** (1 + 1/2^32) = y5 * (y5**(1/2^2)) = y5 * fn sqrt
    
    # 2.1 : define cube_root
    cube_value = y6 # ‡∏Ñ‡πà‡∏≤‡∏™‡∏∏‡∏î‡∏ó‡πâ‡∏≤‡∏¢‡∏´‡∏•‡∏±‡∏á‡∏à‡∏≤‡∏Å‡πÄ‡∏Ç‡πâ‡∏≤‡∏™‡∏π‡∏ï‡∏£
             
    # 2.2 : return cube_value
    return cube_value

# 3 : define fn main : ‡∏£‡∏±‡∏ö input & ‡πÄ‡∏£‡∏µ‡∏¢‡∏Å fn cube_root
def main() :
    q = float(input('Enter Number : '))
    print(cube_root(q)) # ‡πÄ‡∏£‡∏µ‡∏¢‡∏Å fn cube_root ‡∏≠‡∏≠‡∏Å‡∏°‡∏≤
    
# 4 : exec input ‡πÄ‡∏û‡∏∑‡πà‡∏≠ check fn
exec(input()) 
```

### Task 01_08 : Duration (Function) (2 Star)


```python
##### Part 1 : function str2hms ‡∏Ñ‡∏π‡πà‡∏Å‡∏±‡∏ö hms2str

# 1 : ‡∏™‡∏£‡πâ‡∏≤‡∏á fn : ‡πÄ‡πÄ‡∏õ‡∏•‡∏á HH:MM:SS => H (int) , M (int) , S (int)
def str2hms(hms_str):
    t = hms_str.split(':')
    return int(t[0]) , int(t[1]) , int(t[2])

# 2 : ‡∏™‡∏£‡πâ‡∏≤‡∏á fn : ‡πÄ‡πÄ‡∏õ‡∏•‡∏á H (int) , M (int) , S (int) ==> HH:MM:SS
def hms2str(h,m,s) :
    return ('0' + str(h))[-2:] + ':' + \
           ('0' + str(m))[-2:] + ':' + \
           ('0' + str(s))[-2:]
           
# -----------------------------------------------------------------    

##### Part 2 : ‡∏™‡∏£‡πâ‡∏≤‡∏á HMS -> sec ‡∏Ñ‡∏π‡πà‡∏Å‡∏±‡∏ö sec -> HMS

# 3 : fn ‡πÄ‡∏õ‡∏•‡∏µ‡πà‡∏¢‡∏ô H (int) , M (int) , S (int) -> sec_total
def to_sec(h,m,s) :
    sec_total = 3600*h + 60*m + s
    return sec_total

# 4 : fn ‡πÄ‡∏õ‡∏•‡∏µ‡πà‡∏¢‡∏ô sec_total ->  H (int) , M (int) , S (int)
def to_hms(s):
    h = s // 3600
    s = s - (3600*h)
    m = s // 60
    s = s - (60*m)
    s = s
    return h,m,s # int , int , int

# ------------------------------------------------------------------

# 5 : fn ‡πÄ‡∏ó‡∏µ‡∏¢‡∏ö‡∏Ñ‡∏ß‡∏≤‡∏°‡∏ï‡πà‡∏≤‡∏á‡∏≠‡∏≠‡∏Å‡∏°‡∏≤ ==>  ‡∏õ‡πâ‡∏≠‡∏ô HH:MM:SS ‡πÄ‡∏Ç‡πâ‡∏≤‡∏°‡∏≤ return h , m , s
def diff(h1,m1,s1,h2,m2,s2):
    t1 = to_sec(h1,m1,s1)
    t2 = to_sec(h2,m2,s2)
    dt = (86400 + t2 - t1) % 86400 ## Key : ‡∏ö‡∏ß‡∏Å‡∏Ñ‡πà‡∏≤ ‡πÄ‡∏®‡∏© +,- ‡∏ß‡∏¥‡∏ô‡∏≤‡∏ó‡∏µ‡πÑ‡∏õ +1 ‡∏ß‡∏±‡∏ô ==> ‡πÑ‡∏î‡πâ‡∏à‡∏ô.‡∏ß‡∏¥ (~80K) => ‡∏´‡∏≤‡∏£‡πÄ‡∏≠‡∏≤‡πÄ‡∏®‡∏© ‡∏à‡∏ô.‡∏ß‡∏¥‡∏ô‡∏≤‡∏°‡∏µ‡πÉ‡∏ô 1 ‡∏ß‡∏±‡∏ô ‡∏à‡∏∞‡πÑ‡∏î‡πâ 
    ds = to_hms(dt)
    
    return ds # h,m,s

# 6 : fn main() ==> ‡πÄ‡∏£‡∏µ‡∏¢‡∏Å‡∏ó‡∏∏‡∏Å fn ‡∏≠‡∏≠‡∏Å‡∏°‡∏≤
def main() :
    
    # 6.1 : ‡∏£‡∏±‡∏ö input HH:MM:SS ‡πÄ‡∏Ç‡πâ‡∏≤‡∏°‡∏≤ 
    hms_start_str = input('Enter Start HH:MM:SS Time : ') #str
    hms_end_str = input('Enter End HH:MM:SS Time : ') #str
    
    # 6.2 : ‡πÄ‡πÄ‡∏õ‡∏•‡∏á HH:MM:SS ‡πÉ‡∏´‡πâ‡πÄ‡∏õ‡πá‡∏ô H (int) , M (int) , S (int)
    hs , ms , ss = str2hms(hms_start_str)
    he , me , se = str2hms(hms_end_str)
    
    # 6.3 : ‡πÄ‡∏£‡∏µ‡∏¢‡∏Å fn diff => ‡πÑ‡∏î‡πâ‡∏Ñ‡πà‡∏≤ ‡∏à‡∏ô. dh , dm , ds ‡∏≠‡∏≠‡∏Å‡∏°‡∏≤
    diffh , diffm , diffs = diff(hs,ms,ss,he,me,se) # sec
    
    # 6.4 : ‡πÄ‡∏£‡∏µ‡∏¢‡∏Å fn hms2str => ‡πÑ‡∏î‡πâ str ‡πÉ‡∏ô‡∏£‡∏π‡∏õ HH:MM:SS ‡∏≠‡∏≠‡∏Å‡∏°‡∏≤ 
    time_hms = hms2str(diffh , diffm , diffs)
    
    # 6.5 : print ‡∏Ñ‡πà‡∏≤ final ‡∏≠‡∏≠‡∏Å‡∏°‡∏≤ (return ‡∏à‡∏∞‡πÑ‡∏°‡πà‡∏≠‡∏≠‡∏Å‡∏Ñ‡πà‡∏≤ ‡πÄ‡∏û‡∏£‡∏≤‡∏∞‡πÄ‡∏£‡∏≤‡∏£‡∏±‡∏ö input ‡πÉ‡∏ô fn ‡∏ô‡∏µ‡πâ ‡πÄ‡∏•‡∏¢‡∏ï‡πâ‡∏≠‡∏á print ‡∏≠‡∏≠‡∏Å‡∏°‡∏≤)
    print(time_hms)
    
# 7 : exec input ‡πÄ‡∏û‡∏∑‡πà‡∏≠ check fn
exec(input())
```

### Task 01_12 Trigonometry Expression (1 Star)


```python
# 1 : import math + input x
import math
x = float(input('Enter X : '))
x_deg = x # ‡πÉ‡∏´‡πâ‡∏°‡∏∏‡∏°‡πÄ‡∏õ‡πá‡∏ô x degree

# 2 : ‡∏°‡∏≠‡∏á‡∏ó‡∏µ‡∏•‡∏∞‡∏Å‡πâ‡∏≠‡∏ô + ‡πÄ‡πÄ‡∏õ‡∏•‡∏á‡∏°‡∏∏‡∏° deg -> rad (math.cos / math.sin ‡πÉ‡∏ä‡πâ radian)
x_rad = x_deg / 180 * math.pi # x ‡∏°‡∏µ‡∏Ñ‡πà‡∏≤‡πÄ‡∏´‡∏°‡∏∑‡∏≠‡∏ô x_deg

ses = math.sqrt( x**2 + (math.cos(x_rad)**2) )
suan = math.sin(x_rad) + ( (math.cos(x_rad))**3 / (x**2 + 1) )
total = ses / suan

# 3 : output
print('Result of X : ' , round(total,1) )
```

### Task 01_13 : Projectile Motion (1.5 Star)


```python
# 1 : input velocity , theta , time + import math
import math
velocity , theta_deg , time = [float(e) for e in input('Enter u , Œ∏ , t : ').split()]
g = 9.81 # vertical acceleration value

# 2 : transform deg to rad
theta_rad = theta_deg * (math.pi / 180)

# 3 : find h,r by substitute value ( condition : 0 < theta < 90 deg )
if 0 <= theta_deg <= 90 :
    left = (time - (velocity * math.sin(theta_rad)/g) )
    right1 = velocity * math.sin(theta_rad) 
    right2 = (0.5 * g * (time - (3 * velocity * math.sin(theta_rad / g) ) ) )

    height = left * (right1 + right2)
    range = velocity * math.cos(theta_rad) * time
    
    print( round(height*10)/10 , round(range*10)/10 )

else :
    print('Not a Projectile Motion')
```
