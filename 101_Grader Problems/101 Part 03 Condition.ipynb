### Task 03_02 : Intania Major Change (2 Star)


```python
# Key : ‡∏£‡∏±‡∏ö‡∏Ñ‡πà‡∏≤ ‡πÄ‡πÄ‡∏•‡πâ‡∏ß‡πÄ‡∏ä‡πá‡∏Ñ‡∏ó‡∏µ‡∏•‡∏∞‡∏Ñ‡πà‡∏≤ ‡∏ß‡πà‡∏≤‡∏ï‡∏£‡∏á‡∏Å‡∏±‡∏ö‡πÄ‡∏á‡∏∑‡πà‡∏≠‡∏ô‡πÑ‡∏Ç‡πÉ‡∏î

# 1 : ‡∏£‡∏±‡∏ö input -> split ‡πÉ‡∏´‡πâ‡πÄ‡∏õ‡πá‡∏ô list (of str)
stu1 = str(input('Enter first stu_id,gpax,comprog,cal1,cal2 score : ')).split()   # ['stu_id1' , 'gpax1' , 'comprog1' , 'cal1_1', 'cal2_1'] (list (of str))
stu2 = str(input('Enter second stu_id,gpax,comprog,cal1,cal2 score : ')).split()  # ['stu_id2' , 'gpax2' , 'comprog2' , 'cal1_2', 'cal2_2'] (list (of str))

# 2 : ‡πÄ‡∏ó‡∏µ‡∏¢‡∏ö‡πÄ‡πÄ‡∏ï‡πà‡∏•‡∏∞ index ‡πÄ‡πÄ‡∏•‡πâ‡∏ß‡πÄ‡πÄ‡∏ó‡∏ô‡πÄ‡∏Ç‡πâ‡∏≤‡∏ï‡∏±‡∏ß‡πÄ‡πÄ‡∏õ‡∏£ (‡πÄ‡∏û‡∏∑‡πà‡∏≠‡∏Ñ‡∏ß‡∏≤‡∏°‡πÄ‡∏Ç‡πâ‡∏≤‡πÉ‡∏à‡∏á‡πà‡∏≤‡∏¢)

id1 = stu1[0];  gpax1 = float(stu1[1]);     comprog1 = stu1[2];     cal1_1 = stu1[3];     cal2_1 = stu1[4] 
id2 = stu2[0];  gpax2 = float(stu2[1]);     comprog2 = stu2[2];     cal1_2 = stu2[3];     cal2_2 = stu2[4] 

# 2.1 : ‡∏™‡∏£‡πâ‡∏≤‡∏á bool ‡∏ó‡∏µ‡πà‡πÄ‡∏õ‡πá‡∏ô minimum requirement (‡∏Ç‡πâ‡∏≠ 1)

    ## Note : ‡πÄ‡∏õ‡∏£‡∏µ‡∏¢‡∏ö‡πÄ‡∏ó‡∏µ‡∏¢‡∏ö‡∏Ñ‡πà‡∏≤ str ==> 'A' < 'C' = (True) ===> ‡πÄ‡πÄ‡∏õ‡∏•‡∏ß‡πà‡∏≤ ‡πÄ‡∏ß‡∏•‡∏≤‡∏ï‡∏±‡πâ‡∏á‡πÄ‡∏á‡∏∑‡πà‡∏≠‡∏ô‡πÑ‡∏Ç ‡∏à‡∏∞‡∏ï‡πâ‡∏≠‡∏á‡∏ï‡∏±‡πâ‡∏á‡πÉ‡∏´‡πâ user ‡∏Å‡∏£‡∏≠‡∏Å‡∏Ñ‡πà‡∏≤ ‡∏ô‡πâ‡∏≠‡∏¢‡∏Å‡∏ß‡πà‡∏≤ 'C' (‡πÑ‡∏°‡πà‡πÉ‡∏ä‡πà‡πÉ‡∏ô‡πÄ‡∏ä‡∏¥‡∏á‡πÄ‡∏Å‡∏£‡∏î = ‡πÉ‡∏ô‡πÄ‡∏ä‡∏¥‡∏á programming)
minreq1 = (comprog1 == 'A' and cal1_1 <= 'C' and cal2_1 <= 'C') # bool
minreq2 = (comprog2 == 'A' and cal1_2 <= 'C' and cal2_2 <= 'C') # bool

# 2.2 : ‡πÄ‡∏ä‡πá‡∏Ñ condition comprog -> ‡∏ú‡πà‡∏≤‡∏ô min requirement ‡πÄ‡πÄ‡∏•‡πâ‡∏ß

# Case 1 : ‡∏ú‡πà‡∏≤‡∏ô minimum requirement ‡∏ó‡∏±‡πâ‡∏á‡∏Ñ‡∏π‡πà , ‡πÉ‡∏´‡πâ‡πÑ‡∏õ‡πÄ‡∏ä‡πá‡∏Ñ‡∏Ç‡πâ‡∏≠ 2 (‡πÑ‡∏•‡πà‡πÄ‡∏ä‡πá‡∏Ñ‡πÑ‡∏õ‡πÄ‡∏£‡∏∑‡πà‡∏≠‡∏¢‡πÜ)
if (minreq1 and minreq2) : ## if True and True 

    # 01 : ‡πÄ‡∏á‡∏∑‡πà‡∏≠‡∏ô‡πÑ‡∏Ç GPAX => 3 case ‡∏¢‡πà‡∏≠‡∏¢ (if - elif - elif)
    if gpax1 == gpax2 :
        
        # 02 : ‡πÄ‡∏á‡∏∑‡πà‡∏≠‡∏ô‡πÑ‡∏Ç Cal 1 => 3 case ‡∏¢‡πà‡∏≠‡∏¢ (if - elif - elif)
        if cal1_1 == cal1_2 :
            
            # 03 : ‡πÄ‡∏á‡∏∑‡πà‡∏≠‡∏ô‡πÑ‡∏Ç Cal 2 => 3 case (if - elif - elif)
            if cal2_1 == cal2_2 :
                print('Both')
            elif cal2_1 < cal2_2 : # ‡πÄ‡∏ó‡∏µ‡∏¢‡∏östr -> ‡∏Å‡∏•‡∏±‡∏ö‡πÄ‡∏Ñ‡∏£‡∏∑‡πà‡∏≠‡∏á‡∏´‡∏°‡∏≤‡∏¢
                print(id1)
            elif cal2_1 > cal2_2 : # ‡πÄ‡∏ó‡∏µ‡∏¢‡∏östr -> ‡∏Å‡∏•‡∏±‡∏ö‡πÄ‡∏Ñ‡∏£‡∏∑‡πà‡∏≠‡∏á‡∏´‡∏°‡∏≤‡∏¢
                print(id2)
                
        elif cal1_1 < cal1_2 : # ‡πÄ‡∏ó‡∏µ‡∏¢‡∏östr -> ‡∏Å‡∏•‡∏±‡∏ö‡πÄ‡∏Ñ‡∏£‡∏∑‡πà‡∏≠‡∏á‡∏´‡∏°‡∏≤‡∏¢
            print(id1)
            
        elif cal1_1 > cal1_2 : # ‡πÄ‡∏ó‡∏µ‡∏¢‡∏östr -> ‡∏Å‡∏•‡∏±‡∏ö‡πÄ‡∏Ñ‡∏£‡∏∑‡πà‡∏≠‡∏á‡∏´‡∏°‡∏≤‡∏¢
            print(id2)
            
    elif gpax1 > gpax2 :
        print(id1)
        
    elif gpax1 < gpax2 : 
        print(id2)

# Case 2 : ‡∏Ñ‡∏ô‡πÄ‡πÄ‡∏£‡∏Å‡πÑ‡∏°‡πà‡∏ú‡πà‡∏≤‡∏ô ‡∏Ñ‡∏ô‡∏ó‡∏µ‡πà‡∏™‡∏≠‡∏á‡∏ú‡πà‡∏≤‡∏ô => ‡∏ö‡∏±‡∏á‡∏Ñ‡∏±‡∏ö print id ‡∏Ñ‡∏ô‡∏ú‡πà‡∏≤‡∏ô (‡∏Ñ‡∏ô‡∏ó‡∏µ‡πà 2)
elif (not minreq1 and minreq2) : # elif True ==> minreq1 = False , minreq2 = True
    print(id2)

# Case 3 : ‡∏Ñ‡∏ô‡πÄ‡πÄ‡∏£‡∏Å‡∏ú‡πà‡∏≤‡∏ô ‡∏Ñ‡∏ô‡∏ó‡∏µ‡πà‡∏™‡∏≠‡∏á‡πÑ‡∏°‡πà‡∏ú‡πà‡∏≤‡∏ô => ‡∏ö‡∏±‡∏á‡∏Ñ‡∏±‡∏ö print id ‡∏Ñ‡∏ô‡∏ú‡πà‡∏≤‡∏ô (‡∏Ñ‡∏ô‡∏ó‡∏µ‡πà )  
elif (minreq1 and not minreq2) : # elif True ==> minreq1 = True , minreq2 = False
    print(id1)

# Case 4 : ‡πÑ‡∏°‡πà‡∏ú‡πà‡∏≤‡∏ô‡∏ó‡∏±‡πâ‡∏á‡∏Ñ‡∏π‡πà => print None  
else :
    print('None')
```

### Task 03_03 : Gymnastic Score (1 Star)


```python
# Logic : ‡πÄ‡∏ó‡∏µ‡∏¢‡∏ö‡∏Ñ‡πà‡∏≤‡πÑ‡∏õ‡∏ó‡∏∏‡∏Å‡πÜ‡∏Ñ‡πà‡∏≤ -> ‡∏´‡∏≤ min & max -> ‡∏à‡∏±‡∏î‡∏≠‡∏≠‡∏Å ‡πÄ‡πÄ‡∏•‡πâ‡∏ß‡∏≠‡∏µ‡∏Å 2 ‡∏Ñ‡πà‡∏≤‡∏´‡∏≤ avg

gym_score = input().split() # ‡∏´‡∏•‡∏±‡∏á split -> gym_score = ['x1','x2','x3','x4']

x1 = float(gym_score[0]); x2 = float(gym_score[1]); x3 = float(gym_score[2]); x4 = float(gym_score[3])

if (x1 >= x2) and (x1 >= x3) and (x1 >= x4) :
    max_score = x1

elif (x2 >= x1) and (x2 >= x3) and (x2 >= x4) :
    max_score = x2

elif (x3 >= x1) and (x3 >= x2) and (x3 >= x4) :
    max_score = x3
    
elif (x4 >= x1) and (x4 >= x2) and (x4 >= x3) :
    max_score = x4
    
if (x1 <= x2) and (x1 <= x3) and (x1 <= x4) :
    min_score = x1

elif (x2 <= x1) and (x2 <= x3) and (x2 <= x4) :
    min_score = x2

elif (x3 <= x1) and (x3 <= x2) and (x3 <= x4) :
    min_score = x3
    
elif (x4 <= x1) and (x4 <= x2) and (x4 <= x3) :
    min_score = x4

avg_score = ((x1+x2+x3+x4) - (min_score) - (max_score))/2

print( round(avg_score,2) )
    
```

### Solution 03_03 : Gymnastic Score (1 Star) 

```python
# 1 : ‡∏£‡∏±‡∏ö input ‡∏Å‡πà‡∏≠‡∏ô
s1 , s2 , s3 , s4 = input('Enter 4 Referee Points : ').split()
s1 = float(s1);   s2 = float(s2);   s3 = float(s3);   s4 = float(s4); 
# don't forget to change data types

# 2 : check min , max (‡πÄ‡∏õ‡∏•‡∏µ‡πà‡∏¢‡∏ô‡∏Ñ‡πà‡∏≤‡πÄ‡∏≠‡∏≤)
    # 2.1 : cd min
min_score = s1 # ‡πÉ‡∏´‡πâ‡∏Ñ‡πà‡∏≤‡πÄ‡∏£‡∏¥‡πà‡∏°‡∏ï‡πâ‡∏ô s1 = min
if s2 < min_score :
    min_score = s2
if s3 < min_score : # ‡πÄ‡∏ä‡πá‡∏Ñ‡∏ó‡∏∏‡∏Å‡∏Å‡∏£‡∏ì‡∏µ if ‡∏™‡∏∞‡∏î‡∏ß‡∏Å‡∏Å‡∏ß‡πà‡∏≤
    min_score = s3
if s4 < min_score : 
    min_score = s4
    
    # 2.2 : cd max
max_score = s1 
if max_score < s2 :
    max_score = s2
if max_score < s3 : 
    max_score = s3
if max_score < s4 :
    max_score = s4

# 3 : calculate gymnastic score
gym_score = (s1+s2+s3+s4 - min_score - max_score) / 2
print( round (gym_score , 2) )
```

### Task 03_05 : Positive & Negative (1 Star)


```python
# Logic : ‡πÄ‡∏ä‡πá‡∏Ñ positive/negative ‡πÅ‡∏•‡∏∞ ‡πÄ‡∏ä‡πá‡∏Ñ odd/even

# Step 1 : input : ‡∏£‡∏±‡∏ö‡∏Ñ‡πà‡∏≤ input (int)
number = int(input()) #int

# Step 2 : process : ‡πÄ‡∏ä‡πá‡∏Ñ‡∏Ñ‡πà‡∏≤ 2 ‡∏£‡∏≠‡∏ö (condition)

# 2.1  check + or -
if number > 0 : 
    print('positive')
elif number < 0 : # elif ‡πÄ‡∏õ‡πá‡∏ô‡∏Å‡∏≤‡∏£‡πÄ‡∏ä‡πá‡∏Ñ condition ‡∏ï‡πà‡∏≠‡∏à‡∏≤‡∏Å if
    print('negative')
else : # number = 0 
    print('zero')
    
# 2.2  check odd or even
if (number % 2) == 0 : # ‡∏´‡∏≤‡∏£‡πÄ‡∏≠‡∏≤‡πÄ‡∏®‡∏© 2 ‡πÑ‡∏î‡πâ 0 -> ‡∏´‡∏≤‡∏£‡∏•‡∏á‡∏ï‡∏±‡∏ß = even
    print('even')
else : # ‡∏´‡∏≤‡∏£ 2 ‡πÑ‡∏°‡πà‡∏•‡∏á‡∏ï‡∏±‡∏ß -> odd
    print('odd')
```

### Task 03_07 : Abbreviated Number (2 Star)


```python
#### Logic : ‡∏´‡∏≤‡∏£‡∏õ‡∏±‡∏î‡πÄ‡∏®‡∏© 10^3 , 10^6 , 10^9 
# ‡∏ñ‡πâ‡∏≤‡πÄ‡∏õ‡πá‡∏ô‡∏´‡∏•‡∏±‡∏Å‡∏ô‡∏±‡πâ‡∏ô‡πÜ Ex 8K , 6M ==> ‡πÉ‡∏´‡πâ‡πÄ‡∏≠‡∏≤‡∏ó‡∏®‡∏ô‡∏¥‡∏¢‡∏° 1 ‡∏ï‡∏ô.‡∏î‡πâ‡∏ß‡∏¢
# ‡∏ñ‡πâ‡∏≤‡πÄ‡∏Å‡∏¥‡∏ô‡∏´‡∏•‡∏±‡∏Å‡∏ô‡∏±‡πâ‡∏ô‡πÄ‡πÄ‡∏•‡πâ‡∏ß Ex 12K ==> ‡πÑ‡∏°‡πà‡∏ï‡πâ‡∏≠‡∏á‡∏ï‡∏¥‡∏î‡∏ó‡∏®‡∏ô‡∏¥‡∏¢‡∏° ‡∏õ‡∏±‡∏î‡πÄ‡∏õ‡πá‡∏ô K,M,B ‡πÑ‡∏î‡πâ‡πÄ‡∏•‡∏¢

# Step 1 : ‡∏£‡∏±‡∏ö input ‡∏ï‡∏±‡∏ß‡πÄ‡∏•‡∏Ç (int)
number = int(input('Enter Any Integer : ')) #int

# Step 2 : check cd -> ‡∏´‡∏≤‡∏£‡∏õ‡∏±‡∏î‡πÄ‡∏®‡∏©
    
# 2.1 : 0 - 10^3
if 0 <= number < 1000 : # 0 - 999    
    print(number) # print ‡∏ï‡∏≤‡∏°‡∏Ñ‡πà‡∏≤‡∏ô‡∏±‡πâ‡∏ô‡πÄ‡∏•‡∏¢
    
# 2.2 : 10^3 - 10^6 
if 10**3 <= number < 10**6 : # 1K - 999K
    
    # case 1 : ‡∏¢‡∏±‡∏á‡πÑ‡∏°‡πà‡πÄ‡∏Å‡∏¥‡∏ô‡∏´‡∏•‡∏±‡∏Å‡∏ó‡∏µ‡πà‡∏õ‡∏±‡∏î : x 10 -> round -> ‡∏´‡∏≤‡∏£ 10
    if 10**3 <= number < 10**4 :
        k = number / 10**3 
        round_num = (round(10*k)) / 10
        print(str(round_num) + 'K')
    
    # case 2 : ‡πÄ‡∏Å‡∏¥‡∏ô‡∏´‡∏•‡∏±‡∏Å‡∏ó‡∏µ‡πà‡∏õ‡∏±‡∏î‡πÄ‡πÄ‡∏•‡πâ‡∏ß : round ‡∏≠‡∏µ‡∏Å‡∏£‡∏≠‡∏ö‡∏à‡∏≤‡∏Å case 1 (‡πÄ‡∏û‡∏£‡∏≤‡∏∞‡∏ï‡πâ‡∏≠‡∏á‡∏Å‡∏≤‡∏£ int ‡∏´‡∏ô‡πâ‡∏≤ K/M/B)    
    elif 10**4 <= number < 10**6 :
        k = number / 10**3 
        round_num = round((round(10*k)) / 10)
        print(str(round_num) + 'K')

# 2.3 : 10^6 - 10^9         
elif 10**6 <= number < 10**9 : # 100K - 999K
    
    # case 1 : ‡∏¢‡∏±‡∏á‡πÑ‡∏°‡πà‡πÄ‡∏Å‡∏¥‡∏ô‡∏´‡∏•‡∏±‡∏Å‡∏ó‡∏µ‡πà‡∏õ‡∏±‡∏î : x 10 -> round -> ‡∏´‡∏≤‡∏£ 10
    if 10**6 <= number < 10**7 :
        m = number / 10**6 # 
        round_num = (round(10*m)) / 10
        print(str(round_num) + 'M')
        
    # case 2 : ‡πÄ‡∏Å‡∏¥‡∏ô‡∏´‡∏•‡∏±‡∏Å‡∏ó‡∏µ‡πà‡∏õ‡∏±‡∏î‡πÄ‡πÄ‡∏•‡πâ‡∏ß : round ‡∏≠‡∏µ‡∏Å‡∏£‡∏≠‡∏ö‡∏à‡∏≤‡∏Å case 1 (‡πÄ‡∏û‡∏£‡∏≤‡∏∞‡∏ï‡πâ‡∏≠‡∏á‡∏Å‡∏≤‡∏£ int ‡∏´‡∏ô‡πâ‡∏≤ K/M/B)
    elif 10**7 <= number < 10**9 :
        m = number / 10**6 
        round_num = round((round(10*m)) / 10)
        print(str(round_num) + 'M')
           
elif number >= 10**9 : # 100K - 999K
    
    # case 1 : ‡∏¢‡∏±‡∏á‡πÑ‡∏°‡πà‡πÄ‡∏Å‡∏¥‡∏ô‡∏´‡∏•‡∏±‡∏Å‡∏ó‡∏µ‡πà‡∏õ‡∏±‡∏î : x 10 -> round -> ‡∏´‡∏≤‡∏£ 10
    if 10**9 <= number < 10**10 :
        b = number / 10**9 
        round_num = (round(10*b)) / 10
        print(str(round_num) + 'B')
        
    # case 2 : ‡πÄ‡∏Å‡∏¥‡∏ô‡∏´‡∏•‡∏±‡∏Å‡∏ó‡∏µ‡πà‡∏õ‡∏±‡∏î‡πÄ‡πÄ‡∏•‡πâ‡∏ß : round ‡∏≠‡∏µ‡∏Å‡∏£‡∏≠‡∏ö‡∏à‡∏≤‡∏Å case 1 (‡πÄ‡∏û‡∏£‡∏≤‡∏∞‡∏ï‡πâ‡∏≠‡∏á‡∏Å‡∏≤‡∏£ int ‡∏´‡∏ô‡πâ‡∏≤ K/M/B)
    elif number >= 10**10 :
        b = number / 10**9 
        round_num = round((round(10*b)) / 10)
        print(str(round_num) + 'B')
```

```python
nb = input('Enter Number : ')

abbrev10_3x = [' ' , 'K','M','B']

if len(nb) <= 3 :
    print(nb)
    
else : ## ‡∏¢‡∏±‡∏á‡πÑ‡∏°‡πà‡πÄ‡∏ï‡πá‡∏° ‡∏°‡∏µ‡∏õ‡∏±‡∏ç‡∏´‡∏≤‡πÄ‡∏£‡∏∑‡πà‡∏≠‡∏á 1227B --> ‡πÉ‡∏ä‡πâ‡∏ß‡∏¥‡∏ò‡∏µ‡∏î‡πâ‡∏≤‡∏ô‡∏ö‡∏ô‡πÄ‡∏ñ‡∏≠‡∏∞ ‡∏á‡πà‡∏≤‡∏¢‡∏Å‡∏ß‡πà‡∏≤
    
    div = 10 ** (3 * (len(nb) // 3) ) # flooring 10^3x
    
    # case 1 : 10^3 , 10^6 , 10^9 : want 1 decimal : round(10 * float/10^3x) / 10
    
    if len(nb) % 3 == 1 :   # 4 , 7 , 10 , ... digit
        round_nb = round( 10 * (float(nb) / div) ) / 10
        round_str = str(round_nb) + abbrev10_3x [ len(nb) // 3 ]
    
    # case 2 : 10^4 , 10^7 , 10^10 : round to int ==> round( 10 * float/10^3x )
    
    elif len(nb) % 3 == 2 : # 5 , 8 , 11 , ... digit
        round_nb = round( float(nb) / div )
        round_str = str(round_nb) + abbrev10_3x [ len(nb) // 3 ] 
    
    # case 3 : 10^5 , 10^8 , 10^11 : round to int ==> round ( 10 * float/10^3(x-1) )
    
    elif len(nb) % 3 == 0 : # 6 , 9 . 12 , ... digit
        div = 10 ** (3 * (len(nb) // 3 - 1)) #10^3x ‡∏ï‡∏±‡∏ß‡πÄ‡∏î‡∏¥‡∏°
        round_nb = round( float(nb) / div ) 
        round_str = str(round_nb) + abbrev10_3x [ len(nb) // 3 - 1 ] # previous 10^3x
    
    print(round_str)
```

```python
# Main Idea : ‡∏Å‡∏£‡∏≤‡∏ü 3 ‡∏≠‡∏¢‡πà‡∏≤‡∏á (Physical,Emotional,Intellectial) ‡∏°‡∏µ‡∏£‡∏π‡∏õ‡∏Å‡∏£‡∏≤‡∏ü sin -> ‡∏ß‡∏ô‡∏•‡∏π‡∏õ‡∏ó‡∏∏‡∏Å‡πÜ 23,28,33 ‡∏ß‡∏±‡∏ô ‡∏ï‡∏≤‡∏°‡∏•‡∏≥‡∏î‡∏±‡∏ö -> ‡∏´‡∏≤‡∏Ñ‡πà‡∏≤‡∏ó‡∏±‡πâ‡∏á 3

### Logic ‡∏™‡∏±‡∏™‡πÜ ‡πÄ‡∏ô‡πâ‡∏ô‡πÜ ‡∏ï‡∏∂‡∏á‡πÜ
# 1. ‡∏™‡∏£‡πâ‡∏≤‡∏á fn ‡∏Ñ‡∏≥‡∏ô‡∏ß‡∏ì bluezone ‡∏Å‡πà‡∏≠‡∏ô
# 2. ‡∏Ñ‡∏¥‡∏î redzone -> ‡πÉ‡∏ä‡πâ logic ‡πÄ‡∏î‡∏µ‡∏¢‡∏ß‡∏Å‡∏±‡∏ô ‡πÄ‡πÄ‡∏ï‡πà‡∏•‡∏ö‡∏à‡∏≤‡∏Å‡∏ß‡∏±‡∏ô‡∏ó‡∏±‡πâ‡∏á‡∏´‡∏°‡∏î‡∏Ç‡∏≠‡∏á‡∏õ‡∏µ (‡πÄ‡∏£‡∏µ‡∏¢‡∏Å fn bluezone ‡πÄ‡πÄ‡∏ï‡πà‡πÄ‡∏õ‡∏•‡∏µ‡πà‡∏¢‡∏ô parameter)
# 3. blackzone ‡∏î‡∏π‡∏à‡∏ô.‡∏õ‡∏µ‡∏û‡∏≠ (‡πÇ‡∏à‡∏ó‡∏¢‡πå‡∏Å‡∏≥‡∏´‡∏ô‡∏î 365 * ‡∏à‡∏ô.‡∏õ‡∏µ)
# 4. t = 3  ‡πÄ‡∏ß‡∏•‡∏≤‡∏£‡∏ß‡∏°‡∏Å‡∏±‡∏ô + 0 (+1 ‡∏à‡∏≤‡∏Å redzone , -1 ‡∏à‡∏≤‡∏Å bluezone)

## 1 : ‡∏£‡∏±‡∏ö input 3 ‡∏ï‡∏õ. + import math
import math
bd , bm , by_th , d , m , y_th = [int(e) for e in input().split()] # list comp. -> all = int

## 2 : ‡∏™‡∏£‡πâ‡∏≤‡∏á fn ‡∏Ñ‡∏≥‡∏ô‡∏ß‡∏ì ‡∏ä‡πà‡∏ß‡∏á‡∏ü‡πâ‡∏≤ (‡πÄ‡∏£‡∏¥‡πà‡∏°‡∏Ñ‡∏¥‡∏î‡∏ï‡πâ‡∏ô‡∏õ‡∏µ ‡∏ñ‡∏∂‡∏á‡∏ß‡∏±‡∏ô‡∏ó‡∏µ‡πà‡∏™‡∏ô‡πÉ‡∏à -1) (‡πÑ‡∏°‡πà‡πÄ‡∏≠‡∏≤‡∏ß‡∏±‡∏ô‡∏ó‡∏µ‡πà‡∏™‡∏ô‡πÉ‡∏à) 
def bluezone (d , m , y_th) :
    
    # 2.1 : ‡∏™‡∏£‡πâ‡∏≤‡∏á ‡∏ï‡∏õ. & list ‡∏ó‡∏µ‡πà‡∏ï‡πâ‡∏≠‡∏á‡πÉ‡∏ä‡πâ
    sum_day = 0
    noleap = [0,31,28,31,30,31,30,31,31,30,31,30,31]
    leap = [0,31,29,31,30,31,30,31,31,30,31,30,31]

    # 2.2 : ‡∏™‡∏£‡πâ‡∏≤‡∏á‡πÄ‡∏á‡∏∑‡πà‡∏≠‡∏ô‡πÑ‡∏Ç leap year (bool)
    y_en = y_th - 543
    cd_leap1_all = (y_en % 4 == 0 and y_en % 100 == 0 and y_en % 400 == 0) # ‡∏´‡∏≤‡∏£‡∏•‡∏á‡∏´‡∏°‡∏î 4,100,400 -> leap year
    cd_leap2_4 = (y_en % 4 == 0 and y_en % 100 != 0 and y_en % 400 != 0) # ‡∏´‡∏≤‡∏£‡∏•‡∏á‡πÄ‡πÄ‡∏Ñ‡πà 4 -> leap year
    cd_notleap = (y_en % 4 == 0 and y_en % 100 == 0 and y_en % 400 != 0) # ‡∏´‡∏≤‡∏£‡∏•‡∏á‡πÄ‡πÄ‡∏Ñ‡πà 4,100 -> not leap year

    # 2.3 : ‡πÄ‡∏ä‡πá‡∏Ñ cd -> ‡∏ñ‡πâ‡∏≤‡πÄ‡∏õ‡πá‡∏ô leap year ‡∏à‡∏∞‡∏Ñ‡∏≥‡∏ô‡∏ß‡∏ì loop ‡∏Ç‡∏≠‡∏á leap , ‡∏ñ‡πâ‡πà‡∏≤‡πÑ‡∏°‡πà ‡∏Ñ‡∏≥‡∏ô‡∏ß‡∏ì loop ‡∏Ç‡∏≠‡∏á noleap
    if cd_leap1_all or cd_leap2_4 :
        # ‡∏™‡∏£‡πâ‡∏≤‡∏á for loop ‡∏£‡∏ß‡∏° ‡∏à‡∏ô.‡∏ß‡∏±‡∏ô 
        for month in range(1 , m+1 , 1) : # ‡πÄ‡∏£‡∏¥‡πà‡∏°‡πÄ‡∏î‡∏∑‡∏≠‡∏ô‡∏ó‡∏µ‡πà 0 -> ‡∏à‡∏ö‡∏ó‡∏µ‡πà m (‡πÑ‡∏°‡πà‡πÄ‡∏≠‡∏≤ m+1)
            sum_day += (leap[month-1]) # ‡πÑ‡∏°‡πà‡πÄ‡∏≠‡∏≤‡πÄ‡∏î‡∏∑‡∏≠‡∏ô‡∏ï‡∏≠‡∏ô‡∏ô‡∏µ‡πâ ‡∏Ñ‡∏¥‡∏î‡πÄ‡∏â‡∏û‡∏≤‡∏∞‡πÄ‡∏ï‡πá‡∏°‡πÄ‡∏î‡∏∑‡∏≠‡∏ô‡∏Å‡πà‡∏≠‡∏ô
        sum_day += d # ‡πÄ‡∏û‡∏¥‡πà‡∏°‡πÄ‡∏®‡∏©‡∏ß‡∏±‡∏ô‡πÄ‡∏Ç‡πâ‡∏≤‡πÑ‡∏õ
        
    elif cd_notleap or (y_en % 4 != 0) :
        # ‡∏™‡∏£‡πâ‡∏≤‡∏á for loop ‡∏£‡∏ß‡∏° ‡∏à‡∏ô.‡∏ß‡∏±‡∏ô 
        for month in range(1 , m+1 , 1) : # ‡πÄ‡∏£‡∏¥‡πà‡∏°‡πÄ‡∏î‡∏∑‡∏≠‡∏ô‡∏ó‡∏µ‡πà 0 -> ‡∏à‡∏ö‡∏ó‡∏µ‡πà m (‡πÑ‡∏°‡πà‡πÄ‡∏≠‡∏≤ m+1)
            sum_day += (noleap[month-1]) # ‡πÑ‡∏°‡πà‡πÄ‡∏≠‡∏≤‡πÄ‡∏î‡∏∑‡∏≠‡∏ô‡∏ï‡∏≠‡∏ô‡∏ô‡∏µ‡πâ ‡∏Ñ‡∏¥‡∏î‡πÄ‡∏â‡∏û‡∏≤‡∏∞‡πÄ‡∏ï‡πá‡∏°‡πÄ‡∏î‡∏∑‡∏≠‡∏ô‡∏Å‡πà‡∏≠‡∏ô
        sum_day += d # ‡πÄ‡∏û‡∏¥‡πà‡∏°‡πÄ‡∏®‡∏©‡∏ß‡∏±‡∏ô‡πÄ‡∏Ç‡πâ‡∏≤‡πÑ‡∏õ

    # 2.4 : ‡∏≠‡∏¢‡πà‡∏≤‡∏•‡∏∑‡∏° return ‡∏Ñ‡πà‡∏≤ fn 
    return sum_day # ‡πÄ‡∏î‡∏µ‡πã‡∏¢‡∏ß‡πÑ‡∏õ -1 ‡∏ï‡∏≠‡∏ô‡∏ó‡πâ‡∏≤‡∏¢‡∏ó‡∏µ‡πÄ‡∏î‡∏µ‡∏¢‡∏ß

## 3 : ‡∏™‡∏£‡πâ‡∏≤‡∏á cd ‡∏ä‡πà‡∏ß‡∏á‡πÄ‡πÄ‡∏î‡∏á

    # 3.1 : ‡∏™‡∏£‡πâ‡∏≤‡∏á‡πÄ‡∏á‡∏∑‡πà‡∏≠‡∏ô‡πÑ‡∏Ç leap year ‡∏Å‡πà‡∏≠‡∏ô
y_en = y_th - 543
cd_leap1_all = (y_en % 4 == 0 and y_en % 100 == 0 and y_en % 400 == 0) # ‡∏´‡∏≤‡∏£‡∏•‡∏á‡∏´‡∏°‡∏î 4,100,400 -> leap year
cd_leap2_4 = (y_en % 4 == 0 and y_en % 100 != 0 and y_en % 400 != 0) # ‡∏´‡∏≤‡∏£‡∏•‡∏á‡πÄ‡πÄ‡∏Ñ‡πà 4 -> leap year
cd_notleap = (y_en % 4 == 0 and y_en % 100 == 0 and y_en % 400 != 0) # ‡∏´‡∏≤‡∏£‡∏•‡∏á‡πÄ‡πÄ‡∏Ñ‡πà 4,100 -> not leap year

    # 3.2 : ‡πÄ‡∏ä‡πá‡∏Ñ‡πÄ‡∏á‡∏∑‡πà‡∏≠‡∏ô‡πÑ‡∏Ç (Logic ‡πÄ‡∏î‡∏µ‡∏¢‡∏ß‡∏Å‡∏±‡∏ö bluezone ‡πÄ‡πÄ‡∏ï‡πà‡∏Ñ‡∏¥‡∏î‡∏ï‡∏£‡∏á‡∏Ç‡πâ‡∏≤‡∏°)
if cd_leap1_all or cd_leap2_4 :
    redzone = 366 - bluezone(bd,bm,by_th) # ‡∏ô‡∏±‡∏ö‡∏ß‡∏±‡∏ô‡∏ó‡∏µ‡πà‡πÄ‡∏£‡∏≤‡∏™‡∏ô‡πÉ‡∏à‡∏î‡πâ‡∏ß‡∏¢ (+1 ‡∏ó‡∏µ‡∏´‡∏•‡∏±‡∏á)
elif cd_notleap or (y_en % 4 != 0) :
    redzone = 365 - bluezone(bd,bm,by_th) # ‡∏ô‡∏±‡∏ö‡∏ß‡∏±‡∏ô‡∏ó‡∏µ‡πà‡πÄ‡∏£‡∏≤‡∏™‡∏ô‡πÉ‡∏à‡∏î‡πâ‡∏ß‡∏¢ (+1 ‡∏ó‡∏µ‡∏´‡∏•‡∏±‡∏á)

## 4 : ‡∏Ñ‡∏¥‡∏î blackzone
diffyear = y_th - by_th # int
blackzone = 365 * (diffyear - 1) #int

## 5 : ‡∏£‡∏ß‡∏° time -> +1 ‡∏à‡∏≤‡∏Å‡∏™‡∏ô‡πÉ‡∏à redzone , -1 ‡∏à‡∏≤‡∏Å‡πÑ‡∏°‡πà‡πÄ‡∏≠‡∏≤‡∏ß‡∏±‡∏ô bluezone
time = (redzone + 1) + blackzone + (bluezone(d,m,y_th) -1) # int => redzone ‡πÄ‡∏£‡∏µ‡∏¢‡∏Å fn ‡∏Ç‡∏≠‡∏á bluezone ‡∏≠‡∏≠‡∏Å‡∏°‡∏≤‡∏≠‡∏µ‡∏Å‡∏ó‡∏µ

## 6 : ‡πÄ‡∏Ç‡πâ‡∏≤‡∏™‡∏π‡∏ï‡∏£ ‡∏Ñ‡∏≥‡∏ô‡∏ß‡∏ì physical , emotional , intellectual

physical = math.sin( 2*math.pi*time / 23)
emotional = math.sin( 2*math.pi*time / 28)
intellectual = math.sin( 2*math.pi*time / 33)

## 7 : print ‡∏ï‡∏≤‡∏° format (2 ‡∏ï‡∏ô.)
print("{:d} {:.2f} {:.2f} {:.2f}".format(time, physical, emotional, intellectual))
```

### Task 03_14 : Intania Major Change (Function) (2 Star)


```python
# 1 : ‡∏™‡∏£‡πâ‡∏≤‡∏á fn
def choose(stu1,stu2) : ## return ‡∏Ñ‡πà‡∏≤‡∏≠‡∏≠‡∏Å‡∏°‡∏≤‡πÄ‡∏õ‡πá‡∏ô list
    # 1.1 : ‡∏™‡∏£‡πâ‡∏≤‡∏á list ‡πÄ‡∏õ‡∏•‡πà‡∏≤ ‡πÄ‡∏Å‡πá‡∏ö‡πÄ‡∏•‡∏Ç‡∏ô‡∏¥‡∏™‡∏¥‡∏ï ‡πÄ‡∏°‡∏∑‡πà‡∏≠‡∏ú‡∏•‡∏•‡∏±‡∏û‡∏ò‡πå‡∏≠‡∏≠‡∏Å
    list_id = []
    
    # 1.2 : ‡πÅ‡∏ó‡∏ô‡πÄ‡πÄ‡∏ï‡πà‡∏•‡∏∞ index‡πÉ‡∏ô list ‡∏•‡∏á‡πÄ‡πÄ‡∏ï‡πà‡∏•‡∏∞‡∏ï‡∏±‡∏ß‡πÄ‡πÄ‡∏õ‡∏£
    id1 = stu1[0];   gpax1 = float(stu1[1]);   comprog1 = stu1[2];  cal1_1 = stu1[3];  cal2_1 = stu1[4] 
    id2 = stu2[0];   gpax2 = float(stu2[1]);   comprog2 = stu2[2];  cal1_2 = stu2[3];  cal2_2 = stu2[4] 
    
    # 2 : ‡∏™‡∏£‡πâ‡∏≤‡∏á bool ‡∏ó‡∏µ‡πà‡πÄ‡∏õ‡πá‡∏ô minimum requirements ‡∏Å‡πà‡∏≠‡∏ô => 4 case (TT,TF,FT,FF)
    
    minreqcp1 = (comprog1 == 'A' and cal1_1 <= 'C' and cal2_1 <= 'C') # ‡∏£‡∏∞‡∏ß‡∏±‡∏á!! str ‡πÄ‡∏õ‡∏£‡∏µ‡∏¢‡∏ö‡πÄ‡∏ó‡∏µ‡∏¢‡∏ö ‡πÄ‡∏õ‡πá‡∏ô‡∏Ñ‡∏ô‡∏•‡∏∞‡πÄ‡∏á‡∏∑‡πà‡∏≠‡∏ô‡πÑ‡∏Ç‡∏Å‡∏±‡∏ö‡∏Å‡∏≤‡∏£‡πÄ‡∏ó‡∏µ‡∏¢‡∏ö‡πÄ‡∏Å‡∏£‡∏î
    minreqcp2 = (comprog2 == 'A' and cal1_2 <= 'C' and cal2_2 <= 'C')
    
    # case 1 : ‡∏ú‡πà‡∏≤‡∏ô‡∏Ñ‡∏π‡πà -> ‡∏™‡∏£‡πâ‡∏≤‡∏á cd ‡∏ã‡πâ‡∏≠‡∏ô cd (‡πÑ‡∏•‡πà‡πÄ‡∏ä‡πá‡∏Ñ‡∏Ñ‡∏∏‡∏ì‡∏™‡∏°‡∏ö‡∏±‡∏ï‡∏¥‡∏Ç‡πâ‡∏≠ 2)
    if (minreqcp1 and minreqcp2) :
        
        # ‡πÄ‡∏á‡∏∑‡πà‡∏≠‡∏ô‡πÑ‡∏Ç GPAX : 3 case (if - elif - elif)
        if gpax1 == gpax2 :
            
            # ‡πÄ‡∏á‡∏∑‡πà‡∏≠‡∏ô‡πÑ‡∏Ç Cal 1 : 3 case (if - elif - elif)
            if cal1_1 == cal1_2 :
                
                # ‡πÄ‡∏á‡∏∑‡πà‡∏≠‡∏ô‡πÑ‡∏Ç Cal 2 : 3 case (if - elif - elif)
                if cal2_1 == cal2_2 :
                    list_id += [id1 , id2]
                    
                elif cal2_1 < cal2_2 :  # ‡∏£‡∏∞‡∏ß‡∏±‡∏á!! ‡πÄ‡∏õ‡πá‡∏ô‡∏Å‡∏≤‡∏£‡πÄ‡∏õ‡∏£‡∏µ‡∏¢‡∏ö‡πÄ‡∏ó‡∏µ‡∏¢‡∏ö list ‡πÄ‡∏•‡∏¢‡∏Å‡∏•‡∏±‡∏ö‡∏Ñ‡πà‡∏≤‡πÄ‡∏Ñ‡∏£‡∏∑‡πà‡∏≠‡∏á‡∏´‡∏°‡∏≤‡∏¢
                # ‡∏à‡∏∞‡πÑ‡∏î‡πâ‡∏ß‡πà‡∏≤ ‡πÄ‡∏Å‡∏£‡∏î‡∏à‡∏£‡∏¥‡∏á cal2_1 > cal2_2 ==> ‡πÄ‡∏û‡∏¥‡πà‡∏° id1 ‡πÉ‡∏ô list
                    list_id += [id1]                
                
                elif cal2_1 > cal2_2 : # ‡∏£‡∏∞‡∏ß‡∏±‡∏á!! ‡πÄ‡∏õ‡πá‡∏ô‡∏Å‡∏≤‡∏£‡πÄ‡∏õ‡∏£‡∏µ‡∏¢‡∏ö‡πÄ‡∏ó‡∏µ‡∏¢‡∏ö list ‡πÄ‡∏•‡∏¢‡∏Å‡∏•‡∏±‡∏ö‡∏Ñ‡πà‡∏≤‡πÄ‡∏Ñ‡∏£‡∏∑‡πà‡∏≠‡∏á‡∏´‡∏°‡∏≤‡∏¢
                # ‡∏à‡∏∞‡πÑ‡∏î‡πâ‡∏ß‡πà‡∏≤ ‡πÄ‡∏Å‡∏£‡∏î‡∏à‡∏£‡∏¥‡∏á cal2_1 < cal2_2 ==> ‡πÄ‡∏û‡∏¥‡πà‡∏° id2 ‡πÉ‡∏ô list
                    list_id += [id2]
                
            elif cal1_1 < cal1_2 : # ‡∏£‡∏∞‡∏ß‡∏±‡∏á!! ‡πÄ‡∏õ‡πá‡∏ô‡∏Å‡∏≤‡∏£‡πÄ‡∏õ‡∏£‡∏µ‡∏¢‡∏ö‡πÄ‡∏ó‡∏µ‡∏¢‡∏ö list ‡πÄ‡∏•‡∏¢‡∏Å‡∏•‡∏±‡∏ö‡∏Ñ‡πà‡∏≤‡πÄ‡∏Ñ‡∏£‡∏∑‡πà‡∏≠‡∏á‡∏´‡∏°‡∏≤‡∏¢
                # ‡∏à‡∏∞‡πÑ‡∏î‡πâ‡∏ß‡πà‡∏≤ ‡πÄ‡∏Å‡∏£‡∏î‡∏à‡∏£‡∏¥‡∏á cal1_1 > cal1_2 ==> ‡πÄ‡∏û‡∏¥‡πà‡∏° id1 ‡πÉ‡∏ô list
                list_id += [id1]
                
            elif cal1_1 > cal1_2 : # ‡∏£‡∏∞‡∏ß‡∏±‡∏á!! ‡πÄ‡∏õ‡πá‡∏ô‡∏Å‡∏≤‡∏£‡πÄ‡∏õ‡∏£‡∏µ‡∏¢‡∏ö‡πÄ‡∏ó‡∏µ‡∏¢‡∏ö list ‡πÄ‡∏•‡∏¢‡∏Å‡∏•‡∏±‡∏ö‡∏Ñ‡πà‡∏≤‡πÄ‡∏Ñ‡∏£‡∏∑‡πà‡∏≠‡∏á‡∏´‡∏°‡∏≤‡∏¢
                # ‡∏à‡∏∞‡πÑ‡∏î‡πâ‡∏ß‡πà‡∏≤ ‡πÄ‡∏Å‡∏£‡∏î‡∏à‡∏£‡∏¥‡∏á cal1_1 < cal1_2 ==> ‡πÄ‡∏û‡∏¥‡πà‡∏° id2 ‡πÉ‡∏ô list
                list_id += [id2]
            
        elif gpax1 > gpax2 :
            list_id += [id1]
            
        elif gpax1 < gpax2 :
            list_id += [id2]
        
    # case 2 : ‡∏Ñ‡∏ô‡πÄ‡πÄ‡∏£‡∏Å‡∏ú‡πà‡∏≤‡∏ô ‡∏Ñ‡∏ô‡∏™‡∏≠‡∏á‡πÑ‡∏°‡πà‡∏ú‡πà‡∏≤‡∏ô -> ‡πÄ‡∏û‡∏¥‡πà‡∏°‡∏Ñ‡πà‡∏≤ str id1 ‡πÉ‡∏ô list
    elif (minreqcp1 and not minreqcp2) : # if True : minreqcp1 = True , minreqcp2 = False
        list_id += [id1] # list + list only
        
    # case 3 : ‡∏Ñ‡∏ô‡πÄ‡πÄ‡∏£‡∏Å‡πÑ‡∏°‡πà‡∏ú‡πà‡∏≤‡∏ô ‡∏Ñ‡∏ô‡∏™‡∏≠‡∏á‡∏ú‡πà‡∏≤‡∏ô -> ‡πÄ‡∏û‡∏¥‡πà‡∏°‡∏Ñ‡πà‡∏≤ str id2 ‡πÉ‡∏ô list
    elif (not minreqcp1 and minreqcp2) : # if True : minreqcp1 = False , minreqcp2 = True
        list_id += [id2] # list + list only
        
    # case 4 : ‡πÑ‡∏°‡πà‡∏ú‡πà‡∏≤‡∏ô‡∏Ñ‡∏π‡πà -> ‡∏ï‡∏±‡πâ‡∏á‡πÄ‡∏õ‡πá‡∏ô list ‡πÄ‡∏õ‡∏•‡πà‡∏≤
    else : # minreqcp1 = False , minreqcp2 = False
        list_id = []
        
    # 3 : return list_id
    return list_id

# 4 : exec input
exec(input())
```

### Task 03_16 : Day Of Year (Function) (2 Star)


```python
# 1 : ‡∏™‡∏£‡πâ‡∏≤‡∏á fn ‡∏ó‡∏µ‡πà parameter = d(int) , m(int) , y(int)
def day_of_year(d,m,y) :
    
    # 2 : ‡∏™‡∏£‡πâ‡∏≤‡∏á‡πÄ‡∏á‡∏∑‡πà‡∏≠‡∏ô‡πÑ‡∏Ç leap year ‡∏Å‡πà‡∏≠‡∏ô (‡πÑ‡∏õ‡∏•‡∏ö 543 ‡∏Å‡πà‡∏≠‡∏ô) 
     
    # 2.1 : Key ‡∏Ç‡∏≠‡∏á Leap year : ‡∏ö‡∏±‡∏á‡∏Ñ‡∏±‡∏ö‡∏´‡∏≤‡∏£ 4 ‡∏•‡∏á , ‡πÄ‡πÄ‡∏ï‡πà‡∏ñ‡πâ‡∏≤‡∏´‡∏≤‡∏£ 100 ‡∏•‡∏á ‡∏ï‡πâ‡∏≠‡∏á‡∏´‡∏≤‡∏£ 400 ‡∏•‡∏á ==> ‡πÄ‡πÄ‡∏ó‡∏ô‡∏•‡∏á ‡∏ï‡∏õ. m2
    # ‡πÄ‡πÄ‡∏ï‡πà ‡∏ñ‡πâ‡∏≤‡∏´‡∏≤‡∏£ 400 ‡πÑ‡∏°‡πà‡∏•‡∏á ‡πÑ‡∏°‡πà‡πÄ‡∏õ‡πâ‡∏ô leap year (‡∏´‡∏≤‡∏£ 100 ‡∏•‡∏á)
    y -= 543 # ‡πÄ‡πÄ‡∏õ‡∏•‡∏á‡πÄ‡∏õ‡πá‡∏ô ‡∏Ñ‡∏®
    
    
    if (y % 4) == 0 :
        # ‡πÄ‡∏Ç‡πâ‡∏≤‡∏°‡∏≤‡πÄ‡∏ä‡πá‡∏Ñ 100 & 400 ‡∏ï‡πà‡∏≠ => 3 case (100&400‡∏•‡∏á , 100‡∏•‡∏á400‡πÑ‡∏°‡πà‡∏•‡∏á , 100&400‡πÑ‡∏°‡πà‡∏•‡∏á)
        if (y % 100 == 0) and (y % 400 == 0) :
            m2 = 29
        elif (y % 100 == 0) and (y % 400 != 0) :
            m2 = 28
        elif (y % 100 != 0) and (y % 400 != 0) :
            m2 = 29
    else :
        m2 = 28
    ## Note : ‡∏û‡∏¢‡∏≤‡∏¢‡∏≤‡∏°‡∏´‡∏≤‡∏ß‡∏¥‡∏ò‡∏µ‡∏•‡∏î if 12 ‡∏Ç‡∏±‡πâ‡∏ô‡∏ï‡∏≠‡∏ô‡πÉ‡∏´‡πâ‡∏™‡∏±‡πâ‡∏ô‡∏•‡∏á‡πÄ‡∏ñ‡∏≠‡∏∞ ‡∏ï‡∏≠‡∏ô‡∏™‡∏≠‡∏ö‡πÄ‡∏î‡∏µ‡πã‡∏¢‡∏ß‡∏•‡∏≥‡∏ö‡∏≤‡∏Å      
     
    # 2.2 : ‡∏™‡∏£‡πâ‡∏≤‡∏á list ‡∏à‡∏ô.‡∏ß‡∏±‡∏ô‡∏Ç‡∏≠‡∏á‡πÄ‡πÄ‡∏ï‡πà‡∏•‡∏∞‡πÄ‡∏î‡∏∑‡∏≠‡∏ô
    days = [0,31,m2,31,30,31,30,31,31,30,31,30,31] #index 0 ‡∏Å‡∏±‡∏ô‡πÑ‡∏ß‡πâ ‡πÄ‡∏û‡∏£‡∏≤‡∏∞ index m (‡πÄ‡∏•‡∏Ç‡πÄ‡∏î‡∏∑‡∏≠‡∏ô)
    daypast = 0 # ‡∏Å‡∏≥‡∏´‡∏ô‡∏î ‡∏ï‡∏õ.‡πÄ‡∏£‡∏¥‡πà‡∏°‡∏ï‡πâ‡∏ô
    
    # 2.3 : ‡∏™‡∏£‡πâ‡∏≤‡∏á for loop ‡∏ñ‡∏∂‡∏á range ‡πÄ‡∏î‡∏∑‡∏≠‡∏ô‡∏ó‡∏µ‡πà‡πÄ‡∏£‡∏≤‡∏Å‡∏£‡∏≠‡∏Å (‡∏≠‡∏¢‡πà‡∏≤‡∏û‡∏∂‡πà‡∏á‡πÄ‡∏û‡∏¥‡πà‡∏°‡∏Ñ‡πà‡∏≤ d)
    for i in range(m) :  # ‡πÑ‡∏°‡πà‡πÄ‡∏≠‡∏≤ ‡∏ï‡∏ô. ‡πÄ‡∏î‡∏∑‡∏≠‡∏ô‡∏ó‡∏µ‡πà‡πÄ‡∏£‡∏≤‡∏Å‡∏£‡∏≠‡∏Å ‡πÄ‡∏û‡∏£‡∏≤‡∏∞ ‡∏¢‡∏±‡∏á‡πÑ‡∏°‡πà‡πÄ‡∏ï‡πá‡∏°‡πÄ‡∏î‡∏∑‡∏≠‡∏ô                      
        daypast = daypast + days[i] # ‡∏¢‡∏±‡∏á‡πÑ‡∏°‡πà‡∏ö‡∏ß‡∏Å d , ‡πÄ‡∏î‡∏µ‡πã‡∏¢‡∏ß‡πÑ‡∏ß‡πâ‡∏ö‡∏ß‡∏Å‡∏ï‡∏≠‡∏ô‡∏™‡∏∏‡∏î‡∏ó‡πâ‡∏≤‡∏¢
    
    # 2.4 : ‡πÄ‡∏û‡∏¥‡πà‡∏°‡∏Ñ‡πà‡∏≤ d ‡πÄ‡∏Ç‡πâ‡∏≤‡πÑ‡∏õ‡πÉ‡∏ô daypast
    daypast += d # ‡πÄ‡∏û‡∏¥‡πà‡∏°‡∏Ñ‡πà‡∏≤‡∏Ç‡∏≠‡∏á‡πÄ‡∏î‡∏∑‡∏≠‡∏ô‡∏ó‡∏µ‡πà m ‡∏î‡πâ‡∏ß‡∏¢‡∏ß‡∏±‡∏ô‡∏ó‡∏µ‡πà‡∏Ç‡∏≠‡∏á‡πÄ‡∏î‡∏∑‡∏≠‡∏ô‡∏ô‡∏±‡πâ‡∏ô
   
    # 3 : return daypast
    return daypast

# 4 : exec input ‡πÄ‡∏û‡∏∑‡πà‡∏≠ print fn
exec(input())
```
